# Recompilación de operacións morfolóxicas

## <font style="color:rgb(50,120,229)">Apertura e Peche (Opening and Closing)</font>

Como se comentou en clase, podemos combinar operacións de erosión e dilatación para realizar outras con resultados sobre imaxes binarias. Por exemplo, sabemos que a erosión elimina rexión branca e a dilatación a engade. Polo tanto, se queremos **eliminar pequenas rexións brancas**, deberiamos realizar unha **erosión seguida dunha dilatación** para eliminar as manchas brancas máis pequenas, mentres que as máis grandes quedan inalteradas. Igualmente, podemos **eliminar manchas negras usando a dilatación seguida da erosión**.

Por raro que pareza, a estas combinacións de operacións con base na erosión e na dilatación, denomínanse: **Apertura e peche**.

**A apertura** refírese á unha erosión seguida dunha dilatación e o seu fin é eliminar **manchas brancas** e **Peche** denomínase á dilatación seguida da erosión cuxa finalidade é desfacerse de **ocos negros**

Nesta sección, veremos como empregar as operacións de apertura e peche en imaxes binarias. 

## <font style="color:rgb(50,120,229)">Apertura e peche con OpenCV</font>

As operacións de apertura e peche pódense realizar combinando erosión e dilatación. Tamén se pode facer usando funcións OpenCV especiais.

En OpenCV, as operacións de apertura e peche están implementadas coa función **MorphologyEx**.

Para elexir entre aperutura e peche especificaremos un argumento na función [**`MorphologyEx`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html?fbclid=IwAR1GtoDsIv4Fi8o7vrZ8SGb3bb1uiU_Nyt94fc9J2sHKF7FlbDNT1fq-kI0#ga67493776e3ad1a3df63883829375201f). Os argumentos para a apertura e peche son [**`MORPH_OPEN`**] e [**`MORPH_CLOSE`**], respectivamente.

### <font style="color:rgb(8,133,37)">Sintaxe da Función</font>

### <font style="color:rgb(50,120,229)">Apertura</font>

```python:
imageMorphOpened = cv2.morphologyEx( src, cv2.MORPH_OPEN, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]] )
```

### <font style="color:rgb(50,120,229)">Peche</font>

```python:
imageMorphClosed = cv2.morphologyEx( src, cv2.MORPH_CLOSE, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]] )
```

**Parámetros**
- **`src`**	Imaxe orixinal. Número de canles pode ser calquera. O tipo de dato ten que ser CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.
- **`dst`**	Imaxe de destino das mesmas características que a de entrada.
- **`op`**	Tipo de operación morfolóxica
- **`kernel`**	Elemento estrutural. Pode crearse coa función getStructuringElement.
- **`anchor`**	Posición áncora do kernel. Valores negativos indican que o centro é o ancoraxe.
- **`iterations`**	Número de veces que se aplica a operación correspondente (iteracións).
- **`borderType`**	Método de extrapolación.
- **`borderValue`**	Valor do borde en caso de que sexa constante. O valor por defecto ten un significado especial.

**<font style="color:rgb(255,0,0)">Nota:</font>** Nas funcións anteriores, o parámetro 'iteracións' é opcional e se non se menciona, o valor predeterminado é 1. No caso de que necesitemos executar a función de apertura/peche máis dunha vez, especificamos o número de iteracións na chamada á función **MorphologyEx**.

Comentaremos diferentes métodos para realizar operacións de apertura e peche:
1. Usando operacións de dilatación e erosión
1. Empregando a función `MorphologyEx`

## <font style="color:rgb(50,120,229)">Importamos as librerías</font>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
````

## <font style="color:rgb(50,120,229)">Cargamos a imaxe para as opereción de Apertura e Peche</font>

````python
# Lemos a imaxe de entrada
imageName = "../data/opening.png"
image = cv2.imread(imageName, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")

plt.imshow(image)
````

**Obcectivo:** queremos eliminar os círculos brancos pequenos mentres mantemos intactos os círculos grandes

## <font style="color:rgb(50,120,229)">Apertura : Método 1 - Empregamos unha combinación de erosión e dilatación </font>
Neste método, eleximos un núcleo elíptico (elemento estrutural) para realizar a operación de apertura por erosión seguida de dilatación.

````python
# Especificamos o tamaño do kernel
kernelSize = 10

# Creamos o kenel
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*kernelSize+1, 2*kernelSize+1),
                                    (kernelSize, kernelSize))

# Erosión
imEroded = cv2.erode(image, element, iterations=1)
# Dilatación
imOpen = cv2.dilate(imEroded, element, iterations=1)

# Visualizamos o elemento estrutural
plt.imshow(element);plt.title("Elemento estrutural: Elíptico, Tamaño do Kernel: {}".format(kernelSize));plt.show()
````

````python
# Visualización da saída
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe Orixianl")
plt.subplot(132);plt.imshow(imEroded,cmap='gray');plt.title("Despois da operación de erosión")
plt.subplot(133);plt.imshow(imOpen,cmap='gray');plt.title("Despois da operación de dilatación");
````

#### <font style="color:rgb(8,133,37)">Podemos ver que se eliminan todos os círculos pequenos e os máis grandes seguen igual. </font>

## <font style="color:rgb(50,120,229)">Apertura : Método 2 - Empregando a función MorphologyEx</font>

````python
# Creamos o kenel para a peración de apertura
openingSize = 3

# kernel elíptico
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, 
            (2 * openingSize + 1, 2 * openingSize + 1),
            (openingSize,openingSize))
````

Aplicando a erosión seguida da dilatación coa función morphologyEx especificando a etiqueta **`MORPH_OPEN`** (denota a operación de apertura). Nalgúns casos, pode ser necesario implementar unha operación morfolóxica particular máis dunha vez. Neses casos, necesitamos definir o número de iteracións necesarias.

````python
imageMorphOpened = cv2.morphologyEx(image, cv2.MORPH_OPEN, 
                        element,iterations=3)
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe Orixinal")
plt.subplot(122);plt.imshow(imageMorphOpened);plt.title("Despois da Operación de Apertura")
````

## <font style="color:rgb(50,120,229)">Cargamos a imaxe para a operación de Peche</font>

````python
# Cargamos a imaxe de entrada
imageName = "../data/closing.png"
image = cv2.imread(imageName, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
````

**Obxectivo:** Queremos eliminar os círculos pequenos e negros mentres conservamos os máis grandes intactos.

## <font style="color:rgb(50,120,229)">Peche: Método 1 - Usando a combinación de Dilatación e logo Erosión </font>
Neste método, usaremos un núcleo elíptico para realizar a operación de apertura por dilatación seguida da erosión. 

````python
# Tamaño do kernel
kernelSize = 10

# Creamos o kernel
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*kernelSize+1, 2*kernelSize+1),
                                    (kernelSize, kernelSize))

# Dilatación
imDilated = cv2.dilate(image, element)
# Erosión
imClose = cv2.erode(imDilated, element)
````

````python
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe Orixinal")
plt.subplot(132);plt.imshow(imDilated,cmap='gray');plt.title("Despois da operación de Dilatación");
plt.subplot(133);plt.imshow(imClose,cmap='gray');plt.title("Despois da operación de Erosión");
````

#### <font style="color:rgb(8,133,37)">Podemos ver que se eliminan todos os pequenos buratos e os maiores quedan intactos. </font>

## <font style="color:rgb(50,120,229)">Peche: Método 2 - Empregando a función MorphologyEx</font>

````python
# Especificamos o tamaño do kernel para a operación de Peche
closingSize = 10

# Seleccionamos un elemento elíptico
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,
            (2 * closingSize + 1, 2 * closingSize + 1),
            (closingSize,closingSize))
````

Especificamos a operación de peche da imaxe na función morphologyEx especificando o flag **`MORPH_CLOSE`**.

````python
imageMorphClosed = cv2.morphologyEx(image,
                                    cv2.MORPH_CLOSE, element)
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe Orixinal");
plt.subplot(122);plt.imshow(imageMorphClosed,cmap='gray');plt.title("Despois da operación de Peche");
````

## Sombreiros morfolóxicos

### Sombreiro branco e negro

````python
# Cargamos a imaxe
image = cv2.imread("../data/Toyota.jpg", 0)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")

# Construimos un kernel rectagular de (25x5) que coincide coa relación
# ancho/alto dunha matrícula convencional (52x11 cm) e aplicamos un sombreiro negro
# que nos facilitará atopar rexión negras sobro fondos brilantes.
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 5))
blackhat = cv2.morphologyEx(image.copy(), cv2.MORPH_BLACKHAT, rectKernel)

# Igualmente, un sobreiro branco que nos facilitara resaltar
# zonas claras sobre fondos escuros
whitehat = cv2.morphologyEx(image.copy(), cv2.MORPH_TOPHAT, rectKernel)
````

````python
# Visualizamos os resultados
plt.imshow(image);plt.title("Orixinal");plt.show()
plt.imshow(blackhat);plt.title("Sombreiro Negro");plt.show()
plt.imshow(whitehat);plt.title("Sombreiro Branco");plt.show()
````
