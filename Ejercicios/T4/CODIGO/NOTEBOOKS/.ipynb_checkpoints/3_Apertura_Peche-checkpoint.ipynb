# Recompilacin de operacins morfolxicas

## <font style="color:rgb(50,120,229)">Apertura e Peche (Opening and Closing)</font>

Como se comentou en clase, podemos combinar operacins de erosin e dilatacin para realizar outras con resultados sobre imaxes binarias. Por exemplo, sabemos que a erosin elimina rexin branca e a dilatacin a engade. Polo tanto, se queremos **eliminar pequenas rexins brancas**, deberiamos realizar unha **erosin seguida dunha dilatacin** para eliminar as manchas brancas mis pequenas, mentres que as mis grandes quedan inalteradas. Igualmente, podemos **eliminar manchas negras usando a dilatacin seguida da erosin**.

Por raro que pareza, a estas combinacins de operacins con base na erosin e na dilatacin, denomnanse: **Apertura e peche**.

**A apertura** refrese  unha erosin seguida dunha dilatacin e o seu fin  eliminar **manchas brancas** e **Peche** denomnase  dilatacin seguida da erosin cuxa finalidade  desfacerse de **ocos negros**

Nesta seccin, veremos como empregar as operacins de apertura e peche en imaxes binarias. 

## <font style="color:rgb(50,120,229)">Apertura e peche con OpenCV</font>

As operacins de apertura e peche pdense realizar combinando erosin e dilatacin. Tamn se pode facer usando funcins OpenCV especiais.

En OpenCV, as operacins de apertura e peche estn implementadas coa funcin **MorphologyEx**.

Para elexir entre aperutura e peche especificaremos un argumento na funcin [**`MorphologyEx`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html?fbclid=IwAR1GtoDsIv4Fi8o7vrZ8SGb3bb1uiU_Nyt94fc9J2sHKF7FlbDNT1fq-kI0#ga67493776e3ad1a3df63883829375201f). Os argumentos para a apertura e peche son [**`MORPH_OPEN`**] e [**`MORPH_CLOSE`**], respectivamente.

### <font style="color:rgb(8,133,37)">Sintaxe da Funcin</font>

### <font style="color:rgb(50,120,229)">Apertura</font>

```python:
imageMorphOpened = cv2.morphologyEx( src, cv2.MORPH_OPEN, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]] )
```

### <font style="color:rgb(50,120,229)">Peche</font>

```python:
imageMorphClosed = cv2.morphologyEx( src, cv2.MORPH_CLOSE, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]] )
```

**Parmetros**
- **`src`**	Imaxe orixinal. Nmero de canles pode ser calquera. O tipo de dato ten que ser CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.
- **`dst`**	Imaxe de destino das mesmas caractersticas que a de entrada.
- **`op`**	Tipo de operacin morfolxica
- **`kernel`**	Elemento estrutural. Pode crearse coa funcin getStructuringElement.
- **`anchor`**	Posicin ncora do kernel. Valores negativos indican que o centro  o ancoraxe.
- **`iterations`**	Nmero de veces que se aplica a operacin correspondente (iteracins).
- **`borderType`**	Mtodo de extrapolacin.
- **`borderValue`**	Valor do borde en caso de que sexa constante. O valor por defecto ten un significado especial.

**<font style="color:rgb(255,0,0)">Nota:</font>** Nas funcins anteriores, o parmetro 'iteracins'  opcional e se non se menciona, o valor predeterminado  1. No caso de que necesitemos executar a funcin de apertura/peche mis dunha vez, especificamos o nmero de iteracins na chamada  funcin **MorphologyEx**.

Comentaremos diferentes mtodos para realizar operacins de apertura e peche:
1. Usando operacins de dilatacin e erosin
1. Empregando a funcin `MorphologyEx`

## <font style="color:rgb(50,120,229)">Importamos as libreras</font>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
````

## <font style="color:rgb(50,120,229)">Cargamos a imaxe para as operecin de Apertura e Peche</font>

````python
# Lemos a imaxe de entrada
imageName = "../data/opening.png"
image = cv2.imread(imageName, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")

plt.imshow(image)
````

**Obcectivo:** queremos eliminar os crculos brancos pequenos mentres mantemos intactos os crculos grandes

## <font style="color:rgb(50,120,229)">Apertura : Mtodo 1 - Empregamos unha combinacin de erosin e dilatacin </font>
Neste mtodo, eleximos un ncleo elptico (elemento estrutural) para realizar a operacin de apertura por erosin seguida de dilatacin.

````python
# Especificamos o tamao do kernel
kernelSize = 10

# Creamos o kenel
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*kernelSize+1, 2*kernelSize+1),
                                    (kernelSize, kernelSize))

# Erosin
imEroded = cv2.erode(image, element, iterations=1)
# Dilatacin
imOpen = cv2.dilate(imEroded, element, iterations=1)

# Visualizamos o elemento estrutural
plt.imshow(element);plt.title("Elemento estrutural: Elptico, Tamao do Kernel: {}".format(kernelSize));plt.show()
````

````python
# Visualizacin da sada
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe Orixianl")
plt.subplot(132);plt.imshow(imEroded,cmap='gray');plt.title("Despois da operacin de erosin")
plt.subplot(133);plt.imshow(imOpen,cmap='gray');plt.title("Despois da operacin de dilatacin");
````

#### <font style="color:rgb(8,133,37)">Podemos ver que se eliminan todos os crculos pequenos e os mis grandes seguen igual. </font>

## <font style="color:rgb(50,120,229)">Apertura : Mtodo 2 - Empregando a funcin MorphologyEx</font>

````python
# Creamos o kenel para a peracin de apertura
openingSize = 3

# kernel elptico
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, 
            (2 * openingSize + 1, 2 * openingSize + 1),
            (openingSize,openingSize))
````

Aplicando a erosin seguida da dilatacin coa funcin morphologyEx especificando a etiqueta **`MORPH_OPEN`** (denota a operacin de apertura). Nalgns casos, pode ser necesario implementar unha operacin morfolxica particular mis dunha vez. Neses casos, necesitamos definir o nmero de iteracins necesarias.

````python
imageMorphOpened = cv2.morphologyEx(image, cv2.MORPH_OPEN, 
                        element,iterations=3)
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe Orixinal")
plt.subplot(122);plt.imshow(imageMorphOpened);plt.title("Despois da Operacin de Apertura")
````

## <font style="color:rgb(50,120,229)">Cargamos a imaxe para a operacin de Peche</font>

````python
# Cargamos a imaxe de entrada
imageName = "../data/closing.png"
image = cv2.imread(imageName, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
````

**Obxectivo:** Queremos eliminar os crculos pequenos e negros mentres conservamos os mis grandes intactos.

## <font style="color:rgb(50,120,229)">Peche: Mtodo 1 - Usando a combinacin de Dilatacin e logo Erosin </font>
Neste mtodo, usaremos un ncleo elptico para realizar a operacin de apertura por dilatacin seguida da erosin. 

````python
# Tamao do kernel
kernelSize = 10

# Creamos o kernel
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*kernelSize+1, 2*kernelSize+1),
                                    (kernelSize, kernelSize))

# Dilatacin
imDilated = cv2.dilate(image, element)
# Erosin
imClose = cv2.erode(imDilated, element)
````

````python
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe Orixinal")
plt.subplot(132);plt.imshow(imDilated,cmap='gray');plt.title("Despois da operacin de Dilatacin");
plt.subplot(133);plt.imshow(imClose,cmap='gray');plt.title("Despois da operacin de Erosin");
````

#### <font style="color:rgb(8,133,37)">Podemos ver que se eliminan todos os pequenos buratos e os maiores quedan intactos. </font>

## <font style="color:rgb(50,120,229)">Peche: Mtodo 2 - Empregando a funcin MorphologyEx</font>

````python
# Especificamos o tamao do kernel para a operacin de Peche
closingSize = 10

# Seleccionamos un elemento elptico
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,
            (2 * closingSize + 1, 2 * closingSize + 1),
            (closingSize,closingSize))
````

Especificamos a operacin de peche da imaxe na funcin morphologyEx especificando o flag **`MORPH_CLOSE`**.

````python
imageMorphClosed = cv2.morphologyEx(image,
                                    cv2.MORPH_CLOSE, element)
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe Orixinal");
plt.subplot(122);plt.imshow(imageMorphClosed,cmap='gray');plt.title("Despois da operacin de Peche");
````

## Sombreiros morfolxicos

### Sombreiro branco e negro

````python
# Cargamos a imaxe
image = cv2.imread("../data/Toyota.jpg", 0)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")

# Construimos un kernel rectagular de (25x5) que coincide coa relacin
# ancho/alto dunha matrcula convencional (52x11 cm) e aplicamos un sombreiro negro
# que nos facilitar atopar rexin negras sobro fondos brilantes.
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 5))
blackhat = cv2.morphologyEx(image.copy(), cv2.MORPH_BLACKHAT, rectKernel)

# Igualmente, un sobreiro branco que nos facilitara resaltar
# zonas claras sobre fondos escuros
whitehat = cv2.morphologyEx(image.copy(), cv2.MORPH_TOPHAT, rectKernel)
````

````python
# Visualizamos os resultados
plt.imshow(image);plt.title("Orixinal");plt.show()
plt.imshow(blackhat);plt.title("Sombreiro Negro");plt.show()
plt.imshow(whitehat);plt.title("Sombreiro Branco");plt.show()
````
