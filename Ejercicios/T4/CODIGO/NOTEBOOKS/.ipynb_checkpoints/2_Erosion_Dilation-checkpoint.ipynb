# <font style="color:rgb(50,120,229)">Operacións Morfolóxicas</font>


Xa comentamos na clase a teoría das operacións morfolóxicas como a erosión e a dilatación. Resumindo: a **dilatación** úsase para combinar ou expandir rexións brancas que poden estar próximas entre si e a **erosión** emprégase para separar ou reducir as rexións brancas

Nesta caderno, amosaremos como usar as operacións de dilatación e erosión dispoñibles en OpenCV e na seguinte sección veremos que está a suceder baixo o capó destas operacións e como implementalas por conta propia.

# <font style="color:rgb(50,120,229)">Dilatación e Erosión en OpenCV</font>

As operacións de dilatación e erosión, en OpenCV, conséguense empregando [**`dilate`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#ga4ff0f3318642c4f469d0e11f242f3b6c) e [**`erode`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb). Abaixo amósamos as descricións destas funcións.

## <font style="color:rgb(50,120,229)">Dilatación</font>

### <font style="color:rgb(8,133,37)">Sintaxe da Función</font>

```python
dst	=	cv.dilate(	src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]	)
```


## <font style="color:rgb(50,120,229)">Erosión</font>

### <font style="color:rgb(8,133,37)">Sintaxe da Función</font>

```python
dst	=	cv.erode(	src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]	)
```

**Parámetros**

Ámbalas dúas funcións levan o mesmo conxunto de argumentos:

- **`src`**	imaxe de entrada; o número de canles pode ser arbitrario, pero a profundidade debería ser de CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.
- **`dst`**	imaxe de saída do mesmo tamaño e tipo que src.
- **`kernel`**	elemento estruturante empregado para a dilatación; se elemenat = Mat (), úsase un elemento de estruturación rectangular de 3 x 3.
- **`anchor`**	posición da áncora dentro do elemento estruturante; o valor predeterminado (-1, -1) significa que a áncora está no centro do kernel.
- **`iterations`**	número de veces que se aplica a dilatación (iterativamente).
- **`borderType`**	método de extrapolación de píxeles.
- **`borderValue`**	valor do borde en caso de que teña un valor constante.

# <font style="color:rgb(50,120,229)">Importamos as librarias</font>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
````

# <font style="color:rgb(50,120,229)">Cargamos a imaxe</font>

````python
# Lemos a imaxe de entreda e comprobamos que existe
imageName = "../data/dilation_example.jpg"
image = cv2.imread(imageName)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
plt.imshow(image)
````

A imaxe cargada é unha imaxe binaria de 4 moedas. Podemos ver que hai algunhas manchas negras nelas, o que pode ser un problema se queremos crear un algoritmo para detectar as moedas automaticamente. Polo que estamos interesados en eliminar esas manchas para que a imaxe final conste só de 4 zonas brancas (blobs) totalmente.

Como vimos anteriormente, a operación de dilatación pode empregarse para crecer a rexión branca nunha imaxe. Este exemplo, convértese nun candidato perfecto para aplicar dilatación e obter unha imaxe mellorada para un posterior procesamento.

Usarémolo para ilustrar a operación de dilatación.

# <font style="color:rgb(50,120,229)">Código para opereción de Dilatación </font>
Comprobaremos dúas formas de realizar a dilatación.

 1. Usando un único núcleo grande
 2. Empregando varias iteracións dun núcleo máis pequeno


## <font style="color:rgb(50,120,229)">Dilatación empregando un único kernel grande </font>

````python
# Construimos un elemento/kernel estruturante para empregalo na dilatación
kSize = (7,7)
kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, kSize)
plt.imshow(kernel1)
````

````python
# Aplicamos a dilatación sobre a imaxe
imageDilated = cv2.dilate(image, kernel1)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe orixinal")
plt.subplot(122);plt.imshow(imageDilated);plt.title("Imaxe dilatada");
````

## <font style="color:rgb(50,120,229)">Dilatación empregando un kernel pequeno iterativamente </font>

````python
# Construimos un elemento/kernel estruturante para empregalo na dilatación
kSize = (3,3)
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, kSize)
plt.imshow(kernel2)
````

````python
# Aplicamos a dilatación sobre a imaxe
imageDilated1 = cv2.dilate(image, kernel2, iterations=1)
imageDilated2 = cv2.dilate(image, kernel2, iterations=2)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe orixinal")
plt.subplot(132);plt.imshow(imageDilated1);plt.title("Imaxe dilatada: 1 iteración");
plt.subplot(133);plt.imshow(imageDilated2);plt.title("maxe dilatada: 2 iteracións");
````

# <font style="color:rgb(50,120,229)">Código para a operación de Erosión</font>

````python
# Cargamos a imaxe
imageName = "../data/erosion_example.jpg"
image = cv2.imread(imageName)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
plt.imshow(image)
````

A imaxe anterior é só a versión invertida da imaxe empregada na sección de dilatación. Podemos ver que hai 2 manchas brancas en dúas moedas, das que queremos desfacernos para que sexan totalmente negras facilitando a súa análise.

Como vimos anteriormente, a operación de erosión pode usarse para crecer a rexión negra nunha imaxe. Empregarémola para ilustrar a operación de erosión.

Semellante á dilatación, pódese empregar calquera dos 2 métodos para a erosión. A continuación móstrase o código usando un único núcleo. 

````python
# Erosionamos a imaxe
imageEroded = cv2.erode(image, kernel1)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe Orixinal")
plt.subplot(122);plt.imshow(imageEroded);plt.title("Imaxe Erosionada");
````

Podemos empregar varios métodos para detectar a localización das moedas e realizar máis análises desta imaxe empregando os conceptos que aprenderemos nos seguintes cadernos de Jupyter.
