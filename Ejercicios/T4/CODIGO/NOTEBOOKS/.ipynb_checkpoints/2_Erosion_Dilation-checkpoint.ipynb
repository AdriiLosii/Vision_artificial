# <font style="color:rgb(50,120,229)">Operacins Morfolxicas</font>


Xa comentamos na clase a teora das operacins morfolxicas como a erosin e a dilatacin. Resumindo: a **dilatacin** sase para combinar ou expandir rexins brancas que poden estar prximas entre si e a **erosin** emprgase para separar ou reducir as rexins brancas

Nesta caderno, amosaremos como usar as operacins de dilatacin e erosin dispoibles en OpenCV e na seguinte seccin veremos que est a suceder baixo o cap destas operacins e como implementalas por conta propia.

# <font style="color:rgb(50,120,229)">Dilatacin e Erosin en OpenCV</font>

As operacins de dilatacin e erosin, en OpenCV, consguense empregando [**`dilate`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#ga4ff0f3318642c4f469d0e11f242f3b6c) e [**`erode`**](https://docs.opencv.org/4.1.0/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb). Abaixo amsamos as descricins destas funcins.

## <font style="color:rgb(50,120,229)">Dilatacin</font>

### <font style="color:rgb(8,133,37)">Sintaxe da Funcin</font>

```python
dst	=	cv.dilate(	src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]	)
```


## <font style="color:rgb(50,120,229)">Erosin</font>

### <font style="color:rgb(8,133,37)">Sintaxe da Funcin</font>

```python
dst	=	cv.erode(	src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]	)
```

**Parmetros**

mbalas das funcins levan o mesmo conxunto de argumentos:

- **`src`**	imaxe de entrada; o nmero de canles pode ser arbitrario, pero a profundidade debera ser de CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.
- **`dst`**	imaxe de sada do mesmo tamao e tipo que src.
- **`kernel`**	elemento estruturante empregado para a dilatacin; se elemenat = Mat (), sase un elemento de estruturacin rectangular de 3 x 3.
- **`anchor`**	posicin da ncora dentro do elemento estruturante; o valor predeterminado (-1, -1) significa que a ncora est no centro do kernel.
- **`iterations`**	nmero de veces que se aplica a dilatacin (iterativamente).
- **`borderType`**	mtodo de extrapolacin de pxeles.
- **`borderValue`**	valor do borde en caso de que tea un valor constante.

# <font style="color:rgb(50,120,229)">Cargamos a imaxe</font>

````python
# Lemos a imaxe de entreda e comprobamos que existe
imageName = "../data/dilation_example.jpg"
image = cv2.imread(imageName)
if image is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
plt.imshow(image)
````

A imaxe cargada  unha imaxe binaria de 4 moedas. Podemos ver que hai algunhas manchas negras nelas, o que pode ser un problema se queremos crear un algoritmo para detectar as moedas automaticamente. Como vimos anteriormente, a operacin de dilatacin pode empregarse para crecer a rexin branca nunha imaxe. Este exemplo, convrtese nun candidato perfecto para aplicar dilatacin e obter unha imaxe mellorada para un posterior procesamento.

Usarmolo para ilustrar a operacin de dilatacin.

# <font style="color:rgb(50,120,229)">Cdigo para operecin de Dilatacin </font>
Comprobaremos das formas de realizar a dilatacin.

 1. Usando un nico ncleo grande
 2. Empregando varias iteracins dun ncleo mis pequeno


## <font style="color:rgb(50,120,229)">Dilatacin empregando un nico kernel grande </font>

````python
# Construimos un elemento/kernel estruturante para empregalo na dilatacin
kSize = (7,7)
kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, kSize)
plt.imshow(kernel1)
````

````python
# Aplicamos a dilatacin sobre a imaxe
imageDilated = cv2.dilate(image, kernel1)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image);plt.title("Imaxe orixinal")
plt.subplot(122);plt.imshow(imageDilated);plt.title("Imaxe dilatada");
````

## <font style="color:rgb(50,120,229)">Dilatacin empregando un kernel pequeno iterativamente </font>

````python
# Construimos un elemento/kernel estruturante para empregalo na dilatacin
kSize = (3,3)
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, kSize)
plt.imshow(kernel2)
````

````python
# Aplicamos a dilatacin sobre a imaxe
imageDilated1 = cv2.dilate(image, kernel2, iterations=1)
imageDilated2 = cv2.dilate(image, kernel2, iterations=2)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(image);plt.title("Imaxe orixinal")
plt.subplot(132);plt.imshow(imageDilated1);plt.title("Imaxe dilatada: 1 iteracin");
plt.subplot(133);plt.imshow(imageDilated2);plt.title("maxe dilatada: 2 iteracins");
````
