# Template Matching Multi-scala usando OpenCV


### Importamos lpaquetes

```python
from matplotlib import pyplot as plt
import numpy as np
import argparse
import imutils
import glob
import cv2
```

### Funcion para visualizar imaxes en Jupyter Notebooks

```python
def plt_imshow(title, image):
	# convertemos a imaxe de BGR a RGB
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
def resize(image, width = None, height = None, inter = cv2.INTER_AREA):
	# iinicializamos as dimensions da imaxe a ser redimensionada
	dim = None
	(h, w) = image.shape[:2]

	# Se tanto width e height son None, enton retornamos a imaxe orixnal
	if width is None and height is None:
		return image

	# Vemos se width e None
	if width is None:
		# Achamos a ratio entre as height e formamos as dimension desexadas
		r = height / float(h)
		dim = (int(w * r), height)

	# Doutra forma height  None
	else:
		# Achamos a ratio entre as e formamos as dimension desexadas
		r = width / float(w)
		dim = (width, int(h * r))

	# redimensionamos a imaxe
	resized = cv2.resize(image, dim, interpolation = inter)

	# devolvemos a imaxe desexada
	return resized
```

### O truco de `cv2.matchTemplate`

```python

# En Jupiter non empregamos argumentos de entrada ao programa e 
# o reemplazamos p0r parmetros "fixos" 
args = {
    "template": "../data/cod_logo.png",
    "images": "../data/imaxes_TM" #,
    #"visualiza":"True" #Para evitar visualizacions intermedias comenta esta variable dentro do diccionario. Lembrate da coma na lia anterior

}
```

```python
# Cargamos a imaxe, pasamola a gris e detectamos bordes
template = cv2.imread(args["template"])
template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
template = cv2.Canny(template, 50, 200)
(tH, tW) = template.shape[:2]
plt_imshow("Template", template)
```

### Resultados do Template Matching multiescala

```python
# Lazo sobre as imaxes onde quero atopar a forma do template
for imagePath in glob.glob(args["images"] + "/*.jpg"):
	# Cargamos a imaxe e pasmola a gris e inicializamos a variable a
	# bookkeeping para manter a pista das rexins matched
	image = cv2.imread(imagePath)
	gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
	found = None

	# Lazo sobre as escalas da imaxe que queremos analizar
	for scale in np.linspace(0.2, 1.0, 20)[::-1]:
		# Redimensionamos a imaxe acorde a escala e mantemos a pista
        # da relacin de tamanho 
		resized = imutils.resize(gray, width = int(gray.shape[1] * scale))
		r = gray.shape[1] / float(resized.shape[1])

		# Se a imaxe ten un tamanho mais pequeno quo o template, rompemos o lazo
		if resized.shape[0] < tH or resized.shape[1] < tW:
			break

		# Detectamos os bordes na imaxe redimensionada e gris e aplicamos
        #o template matchong para atopar a plantilla desexada
		edged = cv2.Canny(resized, 50, 200)
		result = cv2.matchTemplate(edged, template, cv2.TM_CCOEFF)
		(_, maxVal, _, maxLoc) = cv2.minMaxLoc(result)

		# Comprobamos se a iteracin debe ser visualizada (matching a traves das escalas. Quedamos coa mellor!)
		if args.get("visualiza", False):
			# Debuxamos unha caixa arredor da deteccion
			clone = np.dstack([edged, edged, edged])
			cv2.rectangle(clone, (maxLoc[0], maxLoc[1]),
				(maxLoc[0] + tW, maxLoc[1] + tH), (0, 0, 255), 2)
			cv2.imshow("Visualiza", clone)
			cv2.waitKey(0)

		# Se atopamos un novo maximo no valor da correlacion, enton
        # actualizamos a variable bookkeeping
		if found is None or maxVal > found[0]:
			found = (maxVal, maxLoc, r)

	# Desempaquetamos a varible bookkeeping e achamos as coordenadas (x, y)
	# da caixa envlvente sobre a ratio do resize
	(_, maxLoc, r) = found
	(startX, startY) = (int(maxLoc[0] * r), int(maxLoc[1] * r))
	(endX, endY) = (int((maxLoc[0] + tW) * r), int((maxLoc[1] + tH) * r))

	# Debuxamos a caixa e visualizamos
	cv2.rectangle(image, (startX, startY), (endX, endY), (0, 0, 255), 2)
	plt_imshow("Imaxe", image)
```
