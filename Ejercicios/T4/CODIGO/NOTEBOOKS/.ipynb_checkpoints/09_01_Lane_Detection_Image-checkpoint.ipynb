<h1 style="font-size:30px;">Deteccin de carrs de trfico nunha imaxe esttica </h1>

Este caderno axudarache a crear unha canalizacin para a deteccin de carrs usando OpenCV nunha imaxe (ou frame). Este caso sera o que recollera unha cmara (dianteira) a bordo dun coche apuntando a estrada por onde circula o vehculo. O problema ser abordado a travs da transformada de Hough. A alto nivel, a deteccin de carris nunha estrada realiza os seguintes pasos:

* Establecer un limiar para segmentar as lias brancas horizontais da estrada
* Seleccin da rexin de interese
* Deteccin de bordes mediante o detector de Canny
* Atopar as lias mediante a transformada de Hough
* Extrapolar os carrs a partir das lias atopadas
* Debuxar as zonas detectadas na imaxe orixianl de entrada

````python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
````

# 1. Establecemos un limiar para segmentar as lias brancas

O primeiro paso do noso algoritmo  construr un limiar xeral que aille as lias pintadas no asfalto. Anda que o limiar se aplicar a toda a imaxe, posteriormente illaremos a nosa rexin de interese, polo que anda non precisamos preocuparnos polo ceo, as rbores e outros elementos da escena.

````python
# Lemos a imaxe.
img = cv2.imread('../data/test_img1.jpg')
if img is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15, 10))
plt.imshow(img);
````

````python
# Convertemos a gris
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Empregamos un umbral global baseado nos niveis de gris da escena.
threshold = cv2.inRange(gray, 150, 255)

#isualizamos o resultado.
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1); plt.imshow(gray, cmap = 'gray');      plt.title('Imaxe de gris');
plt.subplot(1,2,2); plt.imshow(threshold, cmap = 'gray'); plt.title('Aplicacin do limiar');
````

# 2. Seleccionamos a rexin de inters coa funcin fillPoly()

Sabemos que as lias son trapezoides //.

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

A funcin `fillPoly()` rechea a rea limitada por un ou mis polgonos.

### <font style="color:rgb(8,133,37)">Sintaxe da Funcin</font>
``` python
dst = cv2.fillPoly( img, pts, color[, lineType[, shift[, offset]]] )
```	
`dst`: imaxe de sada das mesmas mensuras e precisin qe `img`.

A funcin precisa **2 argumentos de entrada** e 4 opcionais:

1. `image`: Imaxe de entrada.
2. `pts`: Array de polgonos onde cada un deles est representado como un array de puntos. 
3. `color`: cor do polgono.
4. `lineType`: tipo de lia coa que debuxarn os polgonos.
5. `shift`: nmero de bits nas coordenadas dos vrtices.
6. `offset`: un offset opcional que afectar a todos os puntos do polgono.

### <font style="color:rgb(8,133,37)">Documentacin de OpenCV</font>

[**`fillPoly()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga311160e71d37e3b795324d097cb3a7dc)
                   
<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
#  Enmascaramento da rexin de interese: 
#Seleccionamos os vertices acorde a imaxe de entrada (manual).
roi_vertices = np.array([[[100, 540],
                          [900, 540],
                          [515, 320],
                          [450, 320]]])

# Definimos unha mscara en branco
mask = np.zeros_like(threshold)   

# Canles de cor da imaxe
if len(threshold.shape) > 2:
    channel_count = threshold.shape[2]  # 3 ou 4 dependendo da imaxe.
    ignore_mask_color = (255,) * channel_count
else:
    ignore_mask_color = 255

# Rechemos o polgono.
cv2.fillPoly(mask, roi_vertices, ignore_mask_color)

# Construimos a rexin de interese en funcin da mscara.
roi = cv2.bitwise_and(threshold, mask)

# isualizamos os resultados
plt.figure(figsize = (20, 10))
plt.subplot(1,3,1); plt.imshow(threshold, cmap = 'gray'); plt.title('Limiar orixinal');
plt.subplot(1,3,2); plt.imshow(mask, cmap = 'gray'); plt.title('Mascara Poligonal');
plt.subplot(1,3,3); plt.imshow(roi, cmap = 'gray'); plt.title('ROI illada');
````

# 3. Detectar bordes e rudo suave

A continuacin necesitamos identificar os bordes da imaxe. Podemos empregar o detector de Canny para identificar as lias de carril, seguido dun suavizado gaussiano para mitigar o rudo que queda despois da deteccin de bordes.

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### <font style="color:rgb(8,133,37)">Documentacin de OpenCV</font>

[**`Canny()`**](https://docs.opencv.org/4.5.2/dd/d1a/group__imgproc__feature.html#ga04723e007ed888ddf11d9ba04e2232de)
[**`GaussianBlur()`**](https://docs.opencv.org/4.5.2/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Realizamos a deteccin de bordes
low_threshold = 50
high_threshold = 100
edges = cv2.Canny(roi, low_threshold, high_threshold)

# Suavizamos cunha gaussiana
kernel_size = 3
canny_blur = cv2.GaussianBlur(edges, (kernel_size, kernel_size), 0)

# Visualizamos as imaxes
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1); plt.imshow(edges, cmap = 'gray'); plt.title('Bordes detectados')
plt.subplot(1,2,2); plt.imshow(canny_blur, cmap = 'gray'); plt.title('Bordes suavizados')
````

# 4. Deteccin de lias empregando a funcin HoughLinesP()

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

`HoughLinesP()` atopa segmentos de lia nunha imaxe binaria empregando a transformada de Hough. A funcin implementa o algoritmo probabilstico de transformada de Hough para a deteccin de lias. 


### <font style="color:rgb(8,133,37)">Sintaxe da funcin</font>
``` python
lines = cv2.HoughLinesP( image, rho, theta, threshold[, lines[, minLineLength[, maxLineGap]]] )
```	
`lines`: Vector de sada coas lias detectadas. Cada lia est representada por un vector de 4 elementos (x1, y1, x2, y2), onde (x1, y1) e (x2, y2) son os puntos terminais de cada segmento detectado.

A funcin ten **4 argumentos de entrada requiridos**:

1. `image`: 8-bit, imaxe binaria de entrada.
2. `rho`: resolucin do acumulador de distancia en pxeles.
3. `theta`: Resolucin angular do acumulador en radins.
4. `threshold`: devolve soamente as lias que superan este limiar.
5. `minLineLength`: Lonxitude mnima das lias. Segmentos mis curtos detectados son rexeitados
6. `maxLineGap`: Mximo "gap" tolerado entre puntos dun segmento para poder unilos.

### <font style="color:rgb(8,133,37)">Documentacin de OpenCV</font>

[**`HoughLinesP()`**](https://docs.opencv.org/4.5.2/dd/d1a/group__imgproc__feature.html#ga8618180a5948286384e3b7ca02f6feeb)
[**`line()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
def draw_lines(img, lines, color = [255, 0, 0], thickness = 2):
    """Utilidade para debuxar lias na imaxe."""
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
````

````python
# Parametros da transformada de Hough elexidos acorde a imaxe de entrada.
rho = 1
theta = np.pi / 180
threshold = 50
min_line_len = 10
max_line_gap = 20

lines = cv2.HoughLinesP(
    canny_blur, rho, theta, threshold, minLineLength = min_line_len, maxLineGap = max_line_gap)

# Debuxamos todas as lias sobre a nova imaxe.
hough = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)
draw_lines(hough, lines)

print("Atopadas {} lias, includas na rexin: {}".format(len(lines), lines[0]))
plt.figure(figsize = (15, 10)); plt.imshow(hough);
````

# 5. Separamos os lados e extrapolamos

### <font style='color:rgb(50,120,230)'> 5.1 Funcin para separar as lias da esquerda e da dereita en funcin da sa pendente </font>

Debemos mirar todas os segmentos detectados e determinar se estn contribundo s lias de carril esquerdo ou dereito. Cando as coordenadas y aumentan de valor de arriba a abaixo, unha lia con pendente positiva forma parte do carril dereito e unha pendente negativa significa o carril esquerdo. Polo tanto, se `y1> y2` temos carril esquerdo e se `y1 <y2` temos carril dereito.

````python
def seperate_left_right_lines(lines):
    """Separamos as lias dos carris en funcin da pendente."""
    left_lines = []
    right_lines = []
    if lines is not None:
        for line in lines:
            for x1, y1, x2, y2 in line:
                if y1 > y2: # Pendente negativa = lia carril esquerdo.
                    left_lines.append([x1, y1, x2, y2])
                elif y1 < y2: # Pendente positiva = lia carril dereito.
                    right_lines.append([x1, y1, x2, y2])
    return left_lines, right_lines
````

### <font style='color:rgb(50,120,230)'> 5.2 funcin para achar a media que se usar na extrapolacin de lias  </font>

````python
def cal_avg(values):
    """Achamos o valor promedio das pendentes e cts."""
    if not (type(values) == 'NoneType'):
        if len(values) > 0:
            n = len(values)
        else:
            n = 1
        return sum(values) / n
````

### <font style='color:rgb(50,120,230)'> 5.3 Funcin para extrapolar lias </font>

Como algunhas lias de carril s se recoecen parcialmente. Deberamos extrapolar a lia para acadar a lonxitude completa. Temos que ter en conta as interseccins de borde inferior e superior para atopar a lonxitude total da lia de carril.

````python
def extrapolate_lines(lines, upper_border, lower_border):
    """Extrapolamos lias mantendo en mente que as inteseccins dos bordes inferior e superior."""
    slopes = []
    consts = []
    
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            slope = (y1-y2) / (x1-x2)
            slopes.append(slope)
            c = y1 - slope * x1
            consts.append(c)
    avg_slope = cal_avg(slopes)
    avg_consts = cal_avg(consts)
    
    # Achamos a intersein promedio no borde inferior.
    x_lower_border_intersections = []
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            x_lower_border_intersection = (lower_border - avg_consts) / avg_slope
            x_lower_border_intersections.append(x_lower_border_intersection)
    x_lane_lower_point = int(cal_avg(x_lower_border_intersections))
    
    # Achamos a intersein promedio no borde superior.
    x_upper_border_intersections = []
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            x_upper_border_intersection = (upper_border - avg_consts) / avg_slope
            x_upper_border_intersections.append(x_upper_border_intersection)
    x_lane_upper_point = int(cal_avg(x_upper_border_intersections))
    return [x_lane_lower_point, lower_border, x_lane_upper_point, upper_border]
````

 ### <font style='color:rgb(50,120,230)'> 5.4 Extraemos as lias extrapoladas e as debuxamos </font>

````python
# Definimos os lmites da rexin de interese.
roi_upper_border = 340
roi_lower_border = 540

# Creamos un array en brando para guarda os resultados (colorizados).
lanes_img = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)

# mpregamos a funcin definida arriba para identifiar listas 
#de lias carril man-esquerda e man-dereitra 
lines_left, lines_right = seperate_left_right_lines(lines)

# Empregamos a funcin definida arriba para extraploar listas de lias nun carril recoecido.
lane_left = extrapolate_lines(lines_left, roi_upper_border, roi_lower_border)
lane_right = extrapolate_lines(lines_right, roi_upper_border, roi_lower_border)
draw_lines(lanes_img, [[lane_left]], thickness = 10)
draw_lines(lanes_img, [[lane_right]], thickness = 10)

# Visualizamos os resultados
fig = plt.figure(figsize = (20, 20))
ax = fig.add_subplot(1, 2, 1); plt.imshow(hough); ax.set_title('Antes da extrapolacin')
ax = fig.add_subplot(1, 2, 2); plt.imshow(lanes_img); ax.set_title('Despois da extrapolacin')
plt.show()
````

# 6. Fusionamos os resultados coa orixinal

Como ltimo paso, podemos combinar os carrs detectados coa imaxe orixinal da estrada.

````python
alpha = 0.8
beta = 1.0
gamma = 0.0
image_annotated = cv2.addWeighted(img, alpha, lanes_img, beta, gamma)

# Visualizamos os resultados e salvamos a disco
plt.imshow(image_annotated)
image_annotated = cv2.cvtColor(image_annotated, cv2.COLOR_BGR2RGB)
cv2.imwrite('./carris-image.jpg', image_annotated);
````
