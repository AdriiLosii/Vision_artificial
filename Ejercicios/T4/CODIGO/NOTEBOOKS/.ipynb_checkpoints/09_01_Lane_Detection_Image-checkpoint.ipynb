<h1 style="font-size:30px;">Detección de carrís de tráfico nunha imaxe estática </h1>

Este caderno axudarache a crear unha canalización para a detección de carrís usando OpenCV nunha imaxe (ou frame). Este caso sería o que recollería unha cámara (dianteira) a bordo dun coche apuntando a estrada por onde circula o vehículo. O problema será abordado a través da transformada de Hough. A alto nivel, a detección de carris nunha estrada realiza os seguintes pasos:

* Establecer un limiar para segmentar as liñas brancas horizontais da estrada
* Selección da rexión de interese
* Detección de bordes mediante o detector de Canny
* Atopar as liñas mediante a transformada de Hough
* Extrapolar os carrís a partir das liñas atopadas
* Debuxar as zonas detectadas na imaxe orixianl de entrada

````python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
````

# 1. Establecemos un limiar para segmentar as liñas brancas

O primeiro paso do noso algoritmo é construír un limiar xeral que aille as liñas pintadas no asfalto. Aínda que o limiar se aplicará a toda a imaxe, posteriormente illaremos a nosa rexión de interese, polo que aínda non precisamos preocuparnos polo ceo, as árbores e outros elementos da escena.

````python
# Lemos a imaxe.
img = cv2.imread('../data/test_img1.jpg')
if img is None:
    raise Exception("Non atopo a imaxe no sitio indicado")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15, 10))
plt.imshow(img);
````

````python
# Convertemos a gris
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Empregamos un umbral global baseado nos niveis de gris da escena.
threshold = cv2.inRange(gray, 150, 255)

#isualizamos o resultado.
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1); plt.imshow(gray, cmap = 'gray');      plt.title('Imaxe de gris');
plt.subplot(1,2,2); plt.imshow(threshold, cmap = 'gray'); plt.title('Aplicación do limiar');
````

# 2. Seleccionamos a rexión de interés coa función fillPoly()

Sabemos que as lias son trapezoides //.

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

A función `fillPoly()` rechea a área limitada por un ou máis polígonos.

### <font style="color:rgb(8,133,37)">Sintaxe da Función</font>
``` python
dst = cv2.fillPoly( img, pts, color[, lineType[, shift[, offset]]] )
```	
`dst`: imaxe de saída das mesmas mensuras e precisión qe `img`.

A función precisa **2 argumentos de entrada** e 4 opcionais:

1. `image`: Imaxe de entrada.
2. `pts`: Array de polígonos onde cada un deles está representado como un array de puntos. 
3. `color`: cor do polígono.
4. `lineType`: tipo de liña coa que debuxarán os polígonos.
5. `shift`: número de bits nas coordenadas dos vértices.
6. `offset`: un offset opcional que afectará a todos os puntos do polígono.

### <font style="color:rgb(8,133,37)">Documentación de OpenCV</font>

[**`fillPoly()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga311160e71d37e3b795324d097cb3a7dc)
                   
<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
#  Enmascaramento da rexión de interese: 
#Seleccionamos os vertices acorde a imaxe de entrada (manual).
roi_vertices = np.array([[[100, 540],
                          [900, 540],
                          [515, 320],
                          [450, 320]]])

# Definimos unha máscara en branco
mask = np.zeros_like(threshold)   

# Canles de cor da imaxe
if len(threshold.shape) > 2:
    channel_count = threshold.shape[2]  # 3 ou 4 dependendo da imaxe.
    ignore_mask_color = (255,) * channel_count
else:
    ignore_mask_color = 255

# Rechemos o polígono.
cv2.fillPoly(mask, roi_vertices, ignore_mask_color)

# Construimos a rexión de interese en función da máscara.
roi = cv2.bitwise_and(threshold, mask)

# isualizamos os resultados
plt.figure(figsize = (20, 10))
plt.subplot(1,3,1); plt.imshow(threshold, cmap = 'gray'); plt.title('Limiar orixinal');
plt.subplot(1,3,2); plt.imshow(mask, cmap = 'gray'); plt.title('Mascara Poligonal');
plt.subplot(1,3,3); plt.imshow(roi, cmap = 'gray'); plt.title('ROI illada');
````

# 3. Detectar bordes e ruído suave

A continuación necesitamos identificar os bordes da imaxe. Podemos empregar o detector de Canny para identificar as liñas de carril, seguido dun suavizado gaussiano para mitigar o ruído que queda despois da detección de bordes.

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

### <font style="color:rgb(8,133,37)">Documentación de OpenCV</font>

[**`Canny()`**](https://docs.opencv.org/4.5.2/dd/d1a/group__imgproc__feature.html#ga04723e007ed888ddf11d9ba04e2232de)
[**`GaussianBlur()`**](https://docs.opencv.org/4.5.2/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
# Realizamos a detección de bordes
low_threshold = 50
high_threshold = 100
edges = cv2.Canny(roi, low_threshold, high_threshold)

# Suavizamos cunha gaussiana
kernel_size = 3
canny_blur = cv2.GaussianBlur(edges, (kernel_size, kernel_size), 0)

# Visualizamos as imaxes
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1); plt.imshow(edges, cmap = 'gray'); plt.title('Bordes detectados')
plt.subplot(1,2,2); plt.imshow(canny_blur, cmap = 'gray'); plt.title('Bordes suavizados')
````

# 4. Detección de liñas empregando a función HoughLinesP()

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

`HoughLinesP()` atopa segmentos de liña nunha imaxe binaria empregando a transformada de Hough. A función implementa o algoritmo probabilístico de transformada de Hough para a detección de liñas. 


### <font style="color:rgb(8,133,37)">Sintaxe da función</font>
``` python
lines = cv2.HoughLinesP( image, rho, theta, threshold[, lines[, minLineLength[, maxLineGap]]] )
```	
`lines`: Vector de saída coas liñas detectadas. Cada liña está representada por un vector de 4 elementos (x1, y1, x2, y2), onde (x1, y1) e (x2, y2) son os puntos terminais de cada segmento detectado.

A función ten **4 argumentos de entrada requiridos**:

1. `image`: 8-bit, imaxe binaria de entrada.
2. `rho`: resolución do acumulador de distancia en píxeles.
3. `theta`: Resolución angular do acumulador en radiáns.
4. `threshold`: devolve soamente as liñas que superan este limiar.
5. `minLineLength`: Lonxitude mínima das liñas. Segmentos máis curtos detectados son rexeitados
6. `maxLineGap`: Máximo "gap" tolerado entre puntos dun segmento para poder unilos.

### <font style="color:rgb(8,133,37)">Documentación de OpenCV</font>

[**`HoughLinesP()`**](https://docs.opencv.org/4.5.2/dd/d1a/group__imgproc__feature.html#ga8618180a5948286384e3b7ca02f6feeb)
[**`line()`**](https://docs.opencv.org/4.5.2/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2)

<hr style="border:none; height: 4px; background-color:#D3D3D3" />

````python
def draw_lines(img, lines, color = [255, 0, 0], thickness = 2):
    """Utilidade para debuxar liñas na imaxe."""
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
````

````python
# Parametros da transformada de Hough elexidos acorde a imaxe de entrada.
rho = 1
theta = np.pi / 180
threshold = 50
min_line_len = 10
max_line_gap = 20

lines = cv2.HoughLinesP(
    canny_blur, rho, theta, threshold, minLineLength = min_line_len, maxLineGap = max_line_gap)

# Debuxamos todas as liñas sobre a nova imaxe.
hough = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)
draw_lines(hough, lines)

print("Atopadas {} liñas, incluídas na rexión: {}".format(len(lines), lines[0]))
plt.figure(figsize = (15, 10)); plt.imshow(hough);
````

# 5. Separamos os lados e extrapolamos

### <font style='color:rgb(50,120,230)'> 5.1 Función para separar as liñas da esquerda e da dereita en función da súa pendente </font>

Debemos mirar todas os segmentos detectados e determinar se están contribuíndo ás liñas de carril esquerdo ou dereito. Cando as coordenadas y aumentan de valor de arriba a abaixo, unha liña con pendente positiva forma parte do carril dereito e unha pendente negativa significa o carril esquerdo. Polo tanto, se `y1> y2` temos carril esquerdo e se `y1 <y2` temos carril dereito.

````python
def seperate_left_right_lines(lines):
    """Separamos as liñas dos carris en función da pendente."""
    left_lines = []
    right_lines = []
    if lines is not None:
        for line in lines:
            for x1, y1, x2, y2 in line:
                if y1 > y2: # Pendente negativa = liña carril esquerdo.
                    left_lines.append([x1, y1, x2, y2])
                elif y1 < y2: # Pendente positiva = liña carril dereito.
                    right_lines.append([x1, y1, x2, y2])
    return left_lines, right_lines
````

### <font style='color:rgb(50,120,230)'> 5.2 función para achar a media que se usará na extrapolación de liñas  </font>

````python
def cal_avg(values):
    """Achamos o valor promedio das pendentes e cts."""
    if not (type(values) == 'NoneType'):
        if len(values) > 0:
            n = len(values)
        else:
            n = 1
        return sum(values) / n
````

### <font style='color:rgb(50,120,230)'> 5.3 Función para extrapolar liñas </font>

Como algunhas liñas de carril só se recoñecen parcialmente. Deberíamos extrapolar a liña para acadar a lonxitude completa. Temos que ter en conta as interseccións de borde inferior e superior para atopar a lonxitude total da liña de carril.

````python
def extrapolate_lines(lines, upper_border, lower_border):
    """Extrapolamos liñas mantendo en mente que as inteseccións dos bordes inferior e superior."""
    slopes = []
    consts = []
    
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            slope = (y1-y2) / (x1-x2)
            slopes.append(slope)
            c = y1 - slope * x1
            consts.append(c)
    avg_slope = cal_avg(slopes)
    avg_consts = cal_avg(consts)
    
    # Achamos a interseión promedio no borde inferior.
    x_lower_border_intersections = []
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            x_lower_border_intersection = (lower_border - avg_consts) / avg_slope
            x_lower_border_intersections.append(x_lower_border_intersection)
    x_lane_lower_point = int(cal_avg(x_lower_border_intersections))
    
    # Achamos a interseión promedio no borde superior.
    x_upper_border_intersections = []
    if lines is not None:
        for x1, y1, x2, y2 in lines:
            x_upper_border_intersection = (upper_border - avg_consts) / avg_slope
            x_upper_border_intersections.append(x_upper_border_intersection)
    x_lane_upper_point = int(cal_avg(x_upper_border_intersections))
    return [x_lane_lower_point, lower_border, x_lane_upper_point, upper_border]
````

 ### <font style='color:rgb(50,120,230)'> 5.4 Extraemos as liñas extrapoladas e as debuxamos </font>

````python
# Definimos os límites da rexión de interese.
roi_upper_border = 340
roi_lower_border = 540

# Creamos un array en brando para guarda os resultados (colorizados).
lanes_img = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)

# mpregamos a función definida arriba para identifiar listas 
#de liñas carril man-esquerda e man-dereitra 
lines_left, lines_right = seperate_left_right_lines(lines)

# Empregamos a función definida arriba para extraploar listas de liñas nun carril recoñecido.
lane_left = extrapolate_lines(lines_left, roi_upper_border, roi_lower_border)
lane_right = extrapolate_lines(lines_right, roi_upper_border, roi_lower_border)
draw_lines(lanes_img, [[lane_left]], thickness = 10)
draw_lines(lanes_img, [[lane_right]], thickness = 10)

# Visualizamos os resultados
fig = plt.figure(figsize = (20, 20))
ax = fig.add_subplot(1, 2, 1); plt.imshow(hough); ax.set_title('Antes da extrapolación')
ax = fig.add_subplot(1, 2, 2); plt.imshow(lanes_img); ax.set_title('Despois da extrapolación')
plt.show()
````

# 6. Fusionamos os resultados coa orixinal

Como último paso, podemos combinar os carrís detectados coa imaxe orixinal da estrada.

````python
alpha = 0.8
beta = 1.0
gamma = 0.0
image_annotated = cv2.addWeighted(img, alpha, lanes_img, beta, gamma)

# Visualizamos os resultados e salvamos a disco
plt.imshow(image_annotated)
image_annotated = cv2.cvtColor(image_annotated, cv2.COLOR_BGR2RGB)
cv2.imwrite('./carris-image.jpg', image_annotated);
````
