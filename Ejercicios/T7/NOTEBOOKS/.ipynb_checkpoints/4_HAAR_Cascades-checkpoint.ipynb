# <font style="color:rgb(50,120,229)">Detección de caras en OpenCV</font>

Nesta sección, veremos unha demostración da implementación de detección de caras, sorrisos e como usar ámbalas cascadas para detectar a cara e o sorriso nunha imaxe. O modelo adestrado para a detección de rostros e sorrisos está presente no cartafol de modelos, que se descarga de [aqui](https://github.com/opencv/opencv/tree/master/data/haarcascades). 

O detector de caras é unha contibución do [Prof. Dr. Rainer Leinhart](http://www.multimedia-computing.de/wiki/Prof._Dr._Rainer_Lienhart) e o detctor de sorrisos de [Deniz Suarez](http://visilab.etsii.uclm.es/personas/oscar/oscar.html). Os modelos gárdanse en ficheiros xml. Hai outros detectores de obxectos baseados no Haar cascade en OpenCV que son adestrados por varios investigadores e que se comparten a través de OpenCV, como detector de ollos, detector de caras de perfil, detector de caras de gato e o detector de peóns. 

A función principal utilizada na demostración é a de `detectMultiscale`

### <font style="color:rgb(8,133,37)">Sintaxe da función</font>

```python
cv2.CascadeClassifier.detectMultiScale(image[, scaleFactor[, minNeighbors]])
```

onde,

- **`image`** imaxe de entrada de grises.
- **`objects`** rexión rectángular que inclúe aos obxectos
- **`scaleFactor`** é o parámetro que especifica canto se reduce o tamaño da imaxe en cada escala da pirámide.
- **`minNeighbors`** é un parámetro que especifica cantos veciños debe ter cada rectángulo candidato para conservalo. O número máis alto dá menos falsos positivos.

## <font style="color:rgb(50,120,229)">Detección de caras</font>

In this example, we discuss the effect of the parameters. The effect of `scaleFactor` is mostly related to speed. Lower the value, slower will be the speed. In this example, we check the effect of the `minNeighbors` parameter. As the value of this variable is increased, false positives are decreased.

Neste exemplo, discutimos a influencia dos parámetros. O efecto de `scaleFactor` está principalmente relacionado coa velocidade. Valores baixos, máis lenta será a velocidade. Neste exemplo, comprobamos como afecta o parámetro `minNeighbors`. A medida que aumenta o seu valor, os falsos positivos diminúen.

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

````python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0,6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
````

````python
# cargamos o Haar cascade dende o ficheiro xml.
faceCascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')
faceNeighborsMax = 10
neighborStep = 1
````

````python
# Lemos a imaxe
frame = cv2.imread("../data/images/hillary_clinton.jpg")
frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
````

````python
# Realizamos a detección multiescala
plt.figure(figsize=(18,18))
count = 1
for neigh in range(1, faceNeighborsMax, neighborStep):
    faces = faceCascade.detectMultiScale(frameGray, 1.2, neigh)
    frameClone = np.copy(frame)
    for (x, y, w, h) in faces:
        cv2.rectangle(frameClone, (x, y), 
                      (x + w, y + h), 
                      (255, 0, 0),2)

    cv2.putText(frameClone, 
    "# vecinhos = {}".format(neigh), (10, 50), 
                cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 6)
    
    plt.subplot(3,3,count)
    plt.imshow(frameClone[:,:,::-1])
    count += 1

plt.show()
````

## <font style="color:rgb(50,120,229)">Detección de caras e sorrisos</font>

O efecto de minNeighbors non foi moi pronunciado no exemplo anterior xa que a cara ten características moi especiais. A boca ou o sorriso é moi difícil de detectar sen falsos positivos. Isto ilustrase co exemplo.

````python
# Detectamos a cara e sorriso empregando o cascade
faceCascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')
smileCascade = cv2.CascadeClassifier('../data/models/haarcascade_smile.xml')
smileNeighborsMax = 90
neighborStep = 10

frame = cv2.imread("../data/images/hillary_clinton.jpg")

frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(frameGray, 1.4, 5)
````

````python
# conseguimos o rectangulo da cara a partir da deteccion
for (x, y, w, h) in faces:
    cv2.rectangle(frame, (x, y), 
                  (x + w, y + h), 
                  (255, 0, 0), 2)
    faceRoiGray = frameGray[y: y + h, x: x + w]
    faceRoiOriginal = frame[y: y + h, x: x + w]

count = 1
plt.figure(figsize=(18,18))
# Detectamos o sorriso
for neigh in range(1, smileNeighborsMax, neighborStep):
    smile = smileCascade.detectMultiScale(faceRoiGray, 
                          1.5, neigh)

    frameClone = np.copy(frame)
    faceRoiClone = frameClone[y: y + h, x: x + w]
    for (xx, yy, ww, hh) in smile:
        cv2.rectangle(faceRoiClone, (xx, yy), 
                      (xx + ww, yy + hh), 
                      (0, 255, 0), 2)

    cv2.putText(frameClone, 
              "# Vecinhos = {}".format(neigh), 
              (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1,
              (0, 0, 255), 4)
    plt.subplot(3,3,count)
    plt.imshow(frameClone[:,:,::-1])
    count += 1

plt.show()
````

Os elementos de Haar son elementos rectangulares inspirados en [Haar wavelets](https://en.wikipedia.org/wiki/Haar_wavelet). O valor dunha característica obtense restando a suma de todos os píxeles da rexión branca da da rexión negra.


$$
f = \Sigma(\text{ píxeles na área negra }) − \Sigma( \text{ píxeles na área branca })
$$

Algúns exemplos móstranse no diagrama anterior. A intuición é que estas características poden detectar bordes, liñas e outros patróns na imaxe e son calcualdas a diferentes escalas e diferentes localizacións como se mostra arriba.

Na figura de arriba, cando a característica **A** se calcula na localización que se mostra na primeira imaxe, dará un valor alto xa que a ponte do nariz é máis clara que a zona circundante. Pero se se calcula noutra escala ou localización (como se mostra na terceira imaxe), a característica dará un valor moito menor. Polo tanto, é unha boa característica para detectar estruturas que se asemellan á ponte nasal.

Do mesmo xeito, cando a característica **B** se calcula como na segunda imaxe, pode detectar a rexión dos ollos xa que é máis escura en comparación a que hai debaixo dela. As funcións **C** e **D** pódense usar para detectar outras características e patróns máis complexos.

Hai que ter en conta que só unha única función non é capaz de detectar caras con gran precisión. Pero, cando moitas destas características votan pola presenza ou ausencia de cara, a detección faise moi precisa e robusta.

Para unha xanela de 24 x 24 utilizada para detectar caras, necesitamos calcular o valor da característica en diferentes escalas e posicións, o que supón ~160.000 características. Será computacionalmente custoso atopar todas as características mesmo cando o cálculo delas implique operacións sinxelas como a suma e a resta. Para abordar este problema, os autores utilizaron **Imaxe Integral**, unha idea tomada de traballos anteriores na área de VA e gráficos.


## <font style="color:rgb(50,120,229)">Que é unha imae integral?</font>

Isto é unha [summed-area table](https://en.wikipedia.org/wiki/Summed-area_table) cuxo valor nun píxel ( x, y ) é a suma de todos os valores de píxeles por riba e á esquerda dese píxel. Como veremos máis adiante, unha vez calculada a imaxe integral, o cálculo das características Haar pódese facerce de forma eficiente.

<center><img src="visuais/opcv4face-w7-m4-image1.png"/></center>

Para o píxel ( x, y ) da Imaxe $I$ mostrada, a Imaxe Integral $II$ vén dada por
$$
II( x, y ) = \sum_{\bf{i \le x , j \le y}} I( i, j )
$$

que é a suma de todos os píxeles da rexión sombreada ( arriba e á esquerda do píxel ( x, y ). A fórmula anterior pódese escribir en formato recursivo e calcular de forma eficiente usando a fórmula:

$$
II( x, y ) = I(x,y) + II(x,y−1) + II(x−1,y) − II(x−1,y−1)
$$

O importante a ter en conta é que podemos calcular a Imaxe Integral ( $II$ ) nunha única pasada usando a fórmula anterior. Podemos usalo para calcular a suma dos valores de píxeles de calquera bloque rectangular da imaxe, usando só 3 operacións aritméticas (1 suma e 2 restas).

Supoñamos que queremos atopar a suma de todos os píxeles na rexión rectangular sombreada <u>abcd</u> para a imaxe que se indica a continuación, a suma necesaria usando Imaxe Integral $II$ vén dada por:
$$
\sum_{x,y \in abcd} I(x,y) = II(a) + II(d) − II(c) − II(b)
$$

## <font style="color:rgb(50,120,229)">Selección de características empregando Adaboost</font>


Como se mencionou anteriormente, non todas as funcións son útiles para unha tarefa de clasificación. Polo tanto, proponse un algoritmo de selección de características para que só se teña que calcular un subconxunto no momento da proba. Estas funcións especiais son seleccionadas mediante o algoritmo Adaboost.

O algoritmo de adestramento selecciona as mellores funcións e aprende o clasificador simultaneamente. Usa a noción de clasificadores débiles, que teñen un rendemento lixeiramente mellor que as adiviñas aleatorias. Sexa $h_i$ un clasificador débil, que vén dado por:

$$
h_{i}({\mathbf {x}})={\begin{cases} −1&{\text{if }}f_{i}<\theta _{i}\\ 1&{\text{noutro caso}}\end{cases}}
$$

Onde, $x$ é a subxanela da imaxe en consideración, $f_i$ é a característica calculada, $\theta_i$ é o limiar que se aprende durante o proceso de adestramento.

Múltiples clasificadores débiles combínanse para construír un clasificador forte. O clasificador forte represéntase como unha combinación lineal de clasificadores débiles.
$$
\bf{h(x)} = \bf{sign}(\sum_i (\alpha_i h_i(x))
$$  

onde $h(x)$ é o clasificador forte e $\alpha_i$ son os pesos que indican a contribución de cada clasificador débil ao clasificador final. Apréndense durante o adestramento.

## <font style="color:rgb(50,120,229)">Cascada de clasificadores</font>

Esta é a fase final do gasoduto. As fiestras secundarias pasan por unha cascada de clasificadores que detectan se unha cara está presente ou non. O diagrama de bloques funcionais do mesmo dáse a continuación.

<center><img src="visuais/opcv4face-w7-m4-blockDiagram.png"/></center>

A idea detrás de usar esta cascada de clasificadores é que a maioría das ventás secundarias dunha imaxe non conterán a cara. Entón, por que desperdiciar recursos ao avaliar todas as funcións desa ventá secundaria. Queremos avaliar as características só daquelas subxanelas que son as candidatas máis prometedoras para ter cara.

O primeiro clasificador usa só unha característica. Rexeita con gran precisión aquelas ventás secundarias que non conteñen caras. As subxanelas que non son rexeitadas polos clasificadores iniciais son avaliadas polo clasificador nas seguintes etapas. A medida que avanzamos na cascada, o número de características avaliadas aumenta e, polo tanto, tamén aumenta a complexidade do clasificador. Isto tamén fai que as fases posteriores sexan máis precisas na detección de caras. No traballo orixinal, a cascada tiña 38 etapas e máis de 6000 características.

Por exemplo, a primeira etapa usa só unha característica, se unha cara está presente, detéctaas case o 100% das veces cunha taxa de falsos positivos do 50%. Do mesmo xeito, a segunda etapa usa 10 funcións e ten un FPR do 20%, etc. Obtemos a saída final (cara ou sen cara) da última etapa que ten unha taxa de erro moi baixa.
