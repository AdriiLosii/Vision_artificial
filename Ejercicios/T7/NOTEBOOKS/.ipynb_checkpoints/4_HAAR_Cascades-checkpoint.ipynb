A funcin principal utilizada na demostracin  a de `detectMultiscale`

### <font style="color:rgb(8,133,37)">Sintaxe da funcin</font>

```python
cv2.CascadeClassifier.detectMultiScale(image[, scaleFactor[, minNeighbors]])
```

onde,

- **`image`** imaxe de entrada de grises.
- **`objects`** rexin rectngular que incle aos obxectos
- **`scaleFactor`**  o parmetro que especifica canto se reduce o tamao da imaxe en cada escala da pirmide.
- **`minNeighbors`**  un parmetro que especifica cantos vecios debe ter cada rectngulo candidato para conservalo. O nmero mis alto d menos falsos positivos.

## <font style="color:rgb(50,120,229)">Deteccin de caras</font>

In this example, we discuss the effect of the parameters. The effect of `scaleFactor` is mostly related to speed. Lower the value, slower will be the speed. In this example, we check the effect of the `minNeighbors` parameter. As the value of this variable is increased, false positives are decreased.

Neste exemplo, discutimos a influencia dos parmetros. O efecto de `scaleFactor` est principalmente relacionado coa velocidade. Valores baixos, mis lenta ser a velocidade. Neste exemplo, comprobamos como afecta o parmetro `minNeighbors`. A medida que aumenta o seu valor, os falsos positivos diminen.

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

````python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0,6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
````

````python
# cargamos o Haar cascade dende o ficheiro xml.
faceCascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')
faceNeighborsMax = 10
neighborStep = 1
````

````python
# Lemos a imaxe
frame = cv2.imread("../data/images/hillary_clinton.jpg")
frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
````

````python
# Realizamos a deteccin multiescala
plt.figure(figsize=(18,18))
count = 1
for neigh in range(1, faceNeighborsMax, neighborStep):
    faces = faceCascade.detectMultiScale(frameGray, 1.2, neigh)
    frameClone = np.copy(frame)
    for (x, y, w, h) in faces:
        cv2.rectangle(frameClone, (x, y), 
                      (x + w, y + h), 
                      (255, 0, 0),2)

    cv2.putText(frameClone, 
    "# vecinhos = {}".format(neigh), (10, 50), 
                cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 6)
    
    plt.subplot(3,3,count)
    plt.imshow(frameClone[:,:,::-1])
    count += 1

plt.show()
````

## <font style="color:rgb(50,120,229)">Deteccin de caras e sorrisos</font>

O efecto de minNeighbors non foi moi pronunciado no exemplo anterior xa que a cara ten caractersticas moi especiais. A boca ou o sorriso  moi difcil de detectar sen falsos positivos. Isto ilustrase co exemplo.

````python
# Detectamos a cara e sorriso empregando o cascade
faceCascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')
smileCascade = cv2.CascadeClassifier('../data/models/haarcascade_smile.xml')
smileNeighborsMax = 90
neighborStep = 10

frame = cv2.imread("../data/images/hillary_clinton.jpg")

frameGray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(frameGray, 1.4, 5)
````

````python
# conseguimos o rectangulo da cara a partir da deteccion
for (x, y, w, h) in faces:
    cv2.rectangle(frame, (x, y), 
                  (x + w, y + h), 
                  (255, 0, 0), 2)
    faceRoiGray = frameGray[y: y + h, x: x + w]
    faceRoiOriginal = frame[y: y + h, x: x + w]

count = 1
plt.figure(figsize=(18,18))
# Detectamos o sorriso
for neigh in range(1, smileNeighborsMax, neighborStep):
    smile = smileCascade.detectMultiScale(faceRoiGray, 
                          1.5, neigh)

    frameClone = np.copy(frame)
    faceRoiClone = frameClone[y: y + h, x: x + w]
    for (xx, yy, ww, hh) in smile:
        cv2.rectangle(faceRoiClone, (xx, yy), 
                      (xx + ww, yy + hh), 
                      (0, 255, 0), 2)

    cv2.putText(frameClone, 
              "# Vecinhos = {}".format(neigh), 
              (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1,
              (0, 0, 255), 4)
    plt.subplot(3,3,count)
    plt.imshow(frameClone[:,:,::-1])
    count += 1

plt.show()
````

Os elementos de Haar son elementos rectangulares inspirados en [Haar wavelets](https://en.wikipedia.org/wiki/Haar_wavelet). O valor dunha caracterstica obtense restando a suma de todos os pxeles da rexin branca da da rexin negra.


$$
f = \Sigma(\text{ pxeles na rea negra })  \Sigma( \text{ pxeles na rea branca })
$$

Para unha xanela de 24 x 24 utilizada para detectar caras, necesitamos calcular o valor da caracterstica en diferentes escalas e posicins, o que supn ~160.000 caractersticas. Para abordar este problema, os autores utilizaron **Imaxe Integral**, unha idea tomada de traballos anteriores na rea de VA e grficos.


## <font style="color:rgb(50,120,229)">Que  unha imae integral?</font>

Isto  unha [summed-area table](https://en.wikipedia.org/wiki/Summed-area_table) cuxo valor nun pxel ( x, y )  a suma de todos os valores de pxeles por riba e  esquerda dese pxel. Como veremos mis adiante, unha vez calculada a imaxe integral, o clculo das caractersticas Haar pdese facerce de forma eficiente.

<center><img src="visuais/opcv4face-w7-m4-image1.png"/></center>

Para o pxel ( x, y ) da Imaxe $I$ mostrada, a Imaxe Integral $II$ vn dada por
$$
II( x, y ) = \sum_{\bf{i \le x , j \le y}} I( i, j )
$$

que  a suma de todos os pxeles da rexin sombreada ( arriba e  esquerda do pxel ( x, y ). A frmula anterior pdese escribir en formato recursivo e calcular de forma eficiente usando a frmula:

$$
II( x, y ) = I(x,y) + II(x,y1) + II(x1,y)  II(x1,y1)
$$

O importante a ter en conta  que podemos calcular a Imaxe Integral ( $II$ ) nunha nica pasada usando a frmula anterior. Podemos usalo para calcular a suma dos valores de pxeles de calquera bloque rectangular da imaxe, usando s 3 operacins aritmticas (1 suma e 2 restas).

Supoamos que queremos atopar a suma de todos os pxeles na rexin rectangular sombreada <u>abcd</u> para a imaxe que se indica a continuacin, a suma necesaria usando Imaxe Integral $II$ vn dada por:
$$
\sum_{x,y \in abcd} I(x,y) = II(a) + II(d)  II(c)  II(b)
$$

## <font style="color:rgb(50,120,229)">Seleccin de caractersticas empregando Adaboost</font>


Como se mencionou anteriormente, non todas as funcins son tiles para unha tarefa de clasificacin. Polo tanto, proponse un algoritmo de seleccin de caractersticas para que s se tea que calcular un subconxunto no momento da proba. Estas funcins especiais son seleccionadas mediante o algoritmo Adaboost.

O algoritmo de adestramento selecciona as mellores funcins e aprende o clasificador simultaneamente. Usa a nocin de clasificadores dbiles, que teen un rendemento lixeiramente mellor que as adivias aleatorias. Sexa $h_i$ un clasificador dbil, que vn dado por:

$$
h_{i}({\mathbf {x}})={\begin{cases} 1&{\text{if }}f_{i}<\theta _{i}\\ 1&{\text{noutro caso}}\end{cases}}
$$

Onde, $x$  a subxanela da imaxe en consideracin, $f_i$  a caracterstica calculada, $\theta_i$  o limiar que se aprende durante o proceso de adestramento.

Mltiples clasificadores dbiles combnanse para construr un clasificador forte. O clasificador forte represntase como unha combinacin lineal de clasificadores dbiles.
$$
\bf{h(x)} = \bf{sign}(\sum_i (\alpha_i h_i(x))
$$  

onde $h(x)$  o clasificador forte e $\alpha_i$ son os pesos que indican a contribucin de cada clasificador dbil ao clasificador final. Aprndense durante o adestramento.

## <font style="color:rgb(50,120,229)">Cascada de clasificadores</font>

Esta  a fase final do gasoduto. As fiestras secundarias pasan por unha cascada de clasificadores que detectan se unha cara est presente ou non. O diagrama de bloques funcionais do mesmo dse a continuacin.

<center><img src="visuais/opcv4face-w7-m4-blockDiagram.png"/></center>

A idea detrs de usar esta cascada de clasificadores  que a maiora das vents secundarias dunha imaxe non contern a cara. Entn, por que desperdiciar recursos ao avaliar todas as funcins desa vent secundaria. Queremos avaliar as caractersticas s daquelas subxanelas que son as candidatas mis prometedoras para ter cara.

O primeiro clasificador usa s unha caracterstica. Rexeita con gran precisin aquelas vents secundarias que non conteen caras. As subxanelas que non son rexeitadas polos clasificadores iniciais son avaliadas polo clasificador nas seguintes etapas. A medida que avanzamos na cascada, o nmero de caractersticas avaliadas aumenta e, polo tanto, tamn aumenta a complexidade do clasificador. Isto tamn fai que as fases posteriores sexan mis precisas na deteccin de caras. No traballo orixinal, a cascada tia 38 etapas e mis de 6000 caractersticas.

Por exemplo, a primeira etapa usa s unha caracterstica, se unha cara est presente, detctaas case o 100% das veces cunha taxa de falsos positivos do 50%. Do mesmo xeito, a segunda etapa usa 10 funcins e ten un FPR do 20%, etc. Obtemos a sada final (cara ou sen cara) da ltima etapa que ten unha taxa de erro moi baixa.
