## <font style="color:rgb(50,120,229)">HOG Descriptor in OpenCV</font>

OpenCV implementa HOG usando a clase HOGDescriptor.

**Python** = hog : **cv2.HOGDescriptor**(winSize, blockSize, blockStride, cellSize, nbins, derivAperture, winSigma, histogramNormType, L2HysThreshold, gammaCorrection, nlevels, useSignedGradients)


<center><img src="visuais/hog+svm-1.png"/></center>

&nbsp;

A imaxe da figura anterior, empregarase como referencia para describir os parámetros importantes dados a continuación. Consulta a descrición de HOG no caderno do tema anterior se o precisas.

- **winSize** = Este parámetro establece o tamaño da xanela sobre a que se calcula o descritor. Nos problemas de clasificación de imaxes, axustámola ao tamaño total da imaxe. Non obstante, se queremos detectar peóns, axustaremola a unha ventana de 64x128 que se axusta ben a estruturas do tipo que queremos detectar.


- **blockSize** = O tamaño da caixa azul da imaxe (ver figura anterior). A noción de bloques existe para abordar a variación da iluminación. Un tamaño de bloque grande fai que os cambios locais sexan menos significativos, mentres que un tamaño de bloque máis pequeno pondera máis os cambios locais. Normalmente blockSize establécese en 2 x cellSize.

- **blockStride** = O blockStride determina a superposición entre os bloques veciños e controla o grao de normalización do contraste. Normalmente, un blockStride establécese ao 50% de blockSize.


- **cellSize** = O cellSize é o tamaño dos cadrados verdes (figura anterior). Elíxese en función da escala das características importantes para facer a clasificación. Un cellSize moi pequeno aumentaría o tamaño do vector de características e un moi grande pode non capturar información relevante.


- **nbins** = Establece o número de contedores no histograma de gradientes. Os autores do artigo HOG recomendaran un valor de 9 para capturar gradientes entre 0 e 180 graos en incrementos de 20 graos.


- **derivAperture** = Tamaño do núcleo de Sobel utilizado para o cálculo do gradiente.

- **winSigma** = Segundo o artigo do HOG, é útil "reducir píxeles preto dos bordes do bloque aplicando unha xanela espacial gaussiana a cada píxel antes de acumular votos de orientación nas celas". winSigma é a desviación estándar desta gaussiana. Na práctica, é mellor deixar este parámetro como predeterminado ( -1 ). Ao facelo, winSigma calcúlase automaticamente como: winSigma = ( blockSize.width + blockSize.height ) / 8


- **histogramNormType** = No artigo do HOG, os autores usan catro tipos diferentes de normalización. OpenCV 3.2 implementa só un deses tipos L2Hys. Entón, simplemente usamos o valor predeterminado. L2Hys é simplemente a normalización L2 seguida dun limiar (L2HysThreshold) onde todos os valores superiores a un limiar se recortan a ese valor.

- **L2HysThreshold** = Limiar utilizado na normalización L2Hys. p.ex. Se a norma L2 dun vector é [0,87, 0,43, 0,22], a normalización L2Hys con L2HysThreshold = 0,8 é [0,8, 0,43, 0,22].


- **gammaCorrection** = Booleano que indica se a corrección gamma debe facerse ou non como paso de pre-procesado.

- **nlevels** = Número de niveis de pirámide utilizados durante a detección. Non ten efecto cando se usa o descritor HOG para a clasificación. 

- **signedGradient** = Normalmente os gradientes poden ter calquera orientación entre 0 e 360 graos. Estes gradientes denomínanse gradientes "con signo" en oposición aos gradientes "sen signo" que só toman valores entre 0 e 180 graos. No artigo orixinal do HOG utilizáronse gradientes sen signo para a detección de peóns.


## <font style="color:rgb(50,120,229)">Adestramento dun clasificador baseado en HOG para detectar gafas</font>
Nesta sección, compartimos o código a través dos pasos para construír un clasificador baseado en HOG. O clasificador mira un ventá na imaxe ao redor dos ollos e clasifícao como persoa con gafas ou sen elas.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0,10.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
predictions2Label = {0: "Sen gafas", 1: "Con gafas"}
```

## <font style="color:rgb(50,120,229)">Detección da cara e determinación da rexión dos ollos</font>

Esta función emprega un Haar cascade xa entrenado en detectar caras frontais na imaxe. Unha vez atopada a rexión que contén a cara, determinamos a rexión dos ollo (heuristicamente sabemos que a altura superior e a inferior están a 1/6 e 3/6 da altura da rexión da cara, respectivamente)

```python
#cargamos o haar-cascade entrenado para detectar caras frontais
face_cascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')

def getCroppedEyeRegion(targetImage):
    
    targetImageGray = cv2.cvtColor(targetImage, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(targetImageGray,1.3,5)
    x,y,w,h = faces[0]

    face_roi = targetImage[y:y+h,x:x+w]
    face_height,face_width = face_roi.shape[:2]

    # Aplicamos a formula heuristica para atopar
    # a rexion dos ollos unha vez atopada a rexion da cara
    eyeTop = int(1/6.0*face_height)
    eyeBottom = int(3/6.0*face_height)
    print("Rexion dos ollos entre comprendida entre: {},{}".format(eyeTop,eyeBottom))
    
    eye_roi = face_roi[eyeTop:eyeBottom,:]
    
    # Redimensionamos o tamanho dos ollos ao fixado na base de datos de 96x32
    cropped = cv2.resize(eye_roi,(96, 32), interpolation = cv2.INTER_CUBIC)
    
    return cropped
```

## <font style="color:rgb(50,120,229)">Preparación dos datos</font>
A seguinte función le as imaxes nun directorio especificado. Crea dúas listas para almacenar imaxes de adestramento e test. Tamén crea listas para etiquetas de adestramento e test. O número de imaxes de proba (test) mantense nun 20% do número total de imaxes.

```python
# Path1 as clases 0 (sen gafas) e Path2 a clase 1(con gafas)
path1 = '../data/images/glassesDataset/cropped_withoutGlasses2'
path2 = '../data/images/glassesDataset/cropped_withGlasses2'

```

```python
def getTrainTest(path, class_val, test_fraction = 0.2):
  testData = []
  trainData = []
  trainLabels = []
  testLabels = []
  inputDir = os.path.expanduser(path)

  # Miramos cantas imaxes hai no directorio e atopamos o numero de adestramento e test
  if os.path.isdir(inputDir):
    images = os.listdir(inputDir)
    images.sort()
    nTest = int(len(images) * test_fraction)

  for counter, img in enumerate(images):

    im = cv2.imread(os.path.join(inputDir, img))
    # Engademos nTest mostras ao conxunto de test
    if counter < nTest:              
      testData.append(im)
      testLabels.append(class_val)
    else:   
      # Engadimos nTrain mostras ao conxunto de entrenamento
      trainData.append(im)
      trainLabels.append(class_val)

  return trainData, trainLabels, testData, testLabels
```

## <font style="color:rgb(50,120,229)">Configuración do clasificador SVM</font>

O clasificador SVM iníciase mediante a función **svmInit**. Establecemos os parámetros como:

- Tipo de núcleo
- Gama
- C

```python
def svmInit(C, gamma):
  model = cv2.ml.SVM_create()
  model.setGamma(gamma)
  model.setC(C)
  model.setKernel(cv2.ml.SVM_RBF)
  model.setType(cv2.ml.SVM_C_SVC)

  return model
```

O clasificador SVM adestrase mediante a función svmTrain. Os datos para a formación deben proporcionarse no formato específico que se indica a continuación. Tamén gardamos o modelo aprendido nun ficheiro yml.

```python
def svmTrain(model, samples, responses):
  model.train(samples, cv2.ml.ROW_SAMPLE, responses)
  return model
```

Esta función úsase para obter a saída ou predición do clasificador.

```python
def svmPredict(model, samples):
  return model.predict(samples)[1].ravel()
```

Esta función atopa as predicións para todas as imaxes de proba da base de datos e atopa a precisión.

```python
def svmEvaluate(model, samples, labels):
  predictions = svmPredict(model, samples)
  accuracy = (labels == predictions).mean()
  print('Porcentaxe precision: %.2f %%' % (accuracy * 100))
  return accuracy
```

A seguinte función converte os descritores HoG de matriz a tipo Mat. Isto é necesario para o formato esixido por SVM

```python
def prepareData(data):
  featureVectorLength = len(data[0])
  features = np.float32(data).reshape(-1, featureVectorLength)
  return features
```

## <font style="color:rgb(50,120,229)">Configura o descritor HOG</font>

A seguinte función calcula os descritores HoG para unha determinada array de imaxes. Debemos especificar as imaxes nunha lista.

```python
def computeHOG(hog, data):

  hogData = []
  for image in data:
    hogFeatures = hog.compute(image)
    hogData.append(hogFeatures)

  return hogData
```

O conxunto de datos para o adestramento gárdase no lugar especificado. As imaxes cárganse mediante as funcións descritas anteriormente

Aquí, especificamos os parámetros do descritor HOG.

```python
# Inicializamos os parametros HoG
winSize = (96, 32)
blockSize = (8, 8)
blockStride = (8, 8)
cellSize = (4, 4)
nbins = 9
derivAperture = 0
winSigma = 4.0
histogramNormType = 1
L2HysThreshold = 2.0000000000000001e-01
gammaCorrection = 1
nlevels = 64

hog = cv2.HOGDescriptor(winSize, blockSize, blockStride, cellSize, 
                      nbins,derivAperture, winSigma, 
                      histogramNormType,L2HysThreshold, 
                      gammaCorrection, nlevels, 1)
```

## <font style="color:rgb(50,120,229)">Cargamos os datos nas variables de entrenamento e test</font>

```python
# Collemos as imaxes de entrenamento e test para as súas clases
negTrainImages, negTrainLabels, negTestImages, negTestLabels = \
getTrainTest(path1, 0, .2)
posTrainImages, posTrainLabels, posTestImages, posTestLabels = \
getTrainTest(path2, 1, .2)

# Engade imaxes positivas e negativas para adestrar e test
trainImages = np.concatenate((np.array(negTrainImages), 
                        np.array(posTrainImages)), 
                            axis=0)
testImages = np.concatenate((np.array(negTestImages), 
                          np.array(posTestImages)), 
                          axis=0)

# AEngade etiquetas positivas e negativas para adestrar e test
trainLabels = np.concatenate((np.array(negTrainLabels), 
                          np.array(posTrainLabels)), 
                          axis=0)
testLabels = np.concatenate((np.array(negTestLabels), 
                          np.array(posTestLabels)), 
                          axis=0)
```

## <font style="color:rgb(50,120,229)">Achamos as caracteristicas</font>

Aquí, calculamos os descritores HOG para os datos de adestramento e proba. A continuación, convertemos os datos a un formato necesario para o clasificador SVM


```python
### Cálculo de características para os datos de adestramento e proba  ##
trainHOG = computeHOG(hog, trainImages)
testHOG = computeHOG(hog, testImages)

# Converte os datos HoG en funcións recoñecidas polo modelo SVM
trainFeatures = prepareData(trainHOG)
testFeatures = prepareData(testHOG)
```

## <font style="color:rgb(50,120,229)">Entrenamos o modelo SVM </font>

Aquí, especificamos os parámetros para SVM Classifier e adestramos o modelo usando as caracteristicas obtidas anteriormente.

```python
###########  adestramento SVM  ##############
model = svmInit(C=2.5, gamma=0.02)  # C = 0.1, gamma 10 for 
#collemos un kernel lineal e salvamos o clasificador entrenado
model = svmTrain(model, trainFeatures, trainLabels)
model.save("eyeGlassClassifierModel.yml")
```

Test e achamos a precisión do modelo para as imaxes de test

## <font style="color:rgb(50,120,229)">Avaliar o modelo e achamos a súa precisión</font>

```python
##########  SVM Test  ###############
# Cargaremos o modelo de novo e probaremos o modelo
# Isto é só para explicar como cargar un modelo SVM
# Podes usar o modelo directamente
savedModel = cv2.ml.SVM_load("eyeGlassClassifierModel.yml")

# Atopa a precisión do modelo
# avaliandoo sobre os datos da test
accuracy = svmEvaluate(savedModel, testFeatures, testLabels)
```

## <font style="color:rgb(50,120,229)">Comprobamos visualmente os resultados</font>

Aquí, usamos o modelo adestrado para realizar a clasificación/predición dunha nova imaxe. As funcións anteriores requiren que as imaxes estean nun vector. En primeiro lugar, recortamos a rexión dos ollos da imaxe usando os puntos de referencia dlib. A continuación, convertemos a imaxe recortada nunha matriz e despois calculamos os descritores HOG. A continuación, cargamos o modelo gardado e usamos a función **predict** para obter as predicións.

```python
# Realizamos un test sobre unha imaxe separada
filename = "../data/images/glassesDataset/glasses_4.jpg"
testImage = cv2.imread(filename)
cropped = getCroppedEyeRegion(testImage)
testHOG = computeHOG(hog, np.array([cropped]))
testFeatures = prepareData(testHOG)
predictions = svmPredict(savedModel, testFeatures)
print("Predicion = {}".format(predictions2Label[int(predictions[0])]))
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(testImage[:,:,::-1])
plt.subplot(1,2,2)
plt.imshow(cropped[:,:,::-1])
plt.show()
```
