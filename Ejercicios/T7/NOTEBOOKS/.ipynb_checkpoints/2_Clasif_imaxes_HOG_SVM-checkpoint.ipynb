## <font style="color:rgb(50,120,229)">HOG Descriptor in OpenCV</font>

OpenCV implementa HOG usando a clase HOGDescriptor.

**Python** = hog : **cv2.HOGDescriptor**(winSize, blockSize, blockStride, cellSize, nbins, derivAperture, winSigma, histogramNormType, L2HysThreshold, gammaCorrection, nlevels, useSignedGradients)

&nbsp;

A imaxe da figura anterior, empregarase como referencia para describir os parmetros importantes dados a continuacin. Consulta a descricin de HOG no caderno do tema anterior se o precisas.

- **winSize** = Este parmetro establece o tamao da xanela sobre a que se calcula o descritor. Nos problemas de clasificacin de imaxes, axustmola ao tamao total da imaxe. Non obstante, se queremos detectar pens, axustaremola a unha ventana de 64x128 que se axusta ben a estruturas do tipo que queremos detectar.


- **blockSize** = O tamao da caixa azul da imaxe (ver figura anterior). A nocin de bloques existe para abordar a variacin da iluminacin. Un tamao de bloque grande fai que os cambios locais sexan menos significativos, mentres que un tamao de bloque mis pequeno pondera mis os cambios locais. Normalmente blockSize establcese en 2 x cellSize.

- **blockStride** = O blockStride determina a superposicin entre os bloques vecios e controla o grao de normalizacin do contraste. Normalmente, un blockStride establcese ao 50% de blockSize.


- **cellSize** = O cellSize  o tamao dos cadrados verdes (figura anterior). Elxese en funcin da escala das caractersticas importantes para facer a clasificacin. Un cellSize moi pequeno aumentara o tamao do vector de caractersticas e un moi grande pode non capturar informacin relevante.


- **nbins** = Establece o nmero de contedores no histograma de gradientes. Os autores do artigo HOG recomendaran un valor de 9 para capturar gradientes entre 0 e 180 graos en incrementos de 20 graos.


- **derivAperture** = Tamao do ncleo de Sobel utilizado para o clculo do gradiente.

- **winSigma** = Segundo o artigo do HOG,  til "reducir pxeles preto dos bordes do bloque aplicando unha xanela espacial gaussiana a cada pxel antes de acumular votos de orientacin nas celas". winSigma  a desviacin estndar desta gaussiana. Na prctica,  mellor deixar este parmetro como predeterminado ( -1 ). Ao facelo, winSigma calclase automaticamente como: winSigma = ( blockSize.width + blockSize.height ) / 8


- **histogramNormType** = No artigo do HOG, os autores usan catro tipos diferentes de normalizacin. OpenCV 3.2 implementa s un deses tipos L2Hys. Entn, simplemente usamos o valor predeterminado. L2Hys  simplemente a normalizacin L2 seguida dun limiar (L2HysThreshold) onde todos os valores superiores a un limiar se recortan a ese valor.

- **L2HysThreshold** = Limiar utilizado na normalizacin L2Hys. p.ex. Se a norma L2 dun vector  [0,87, 0,43, 0,22], a normalizacin L2Hys con L2HysThreshold = 0,8  [0,8, 0,43, 0,22].


- **gammaCorrection** = Booleano que indica se a correccin gamma debe facerse ou non como paso de pre-procesado.

- **nlevels** = Nmero de niveis de pirmide utilizados durante a deteccin. Non ten efecto cando se usa o descritor HOG para a clasificacin. 

- **signedGradient** = Normalmente os gradientes poden ter calquera orientacin entre 0 e 360 graos. Estes gradientes denomnanse gradientes "con signo" en oposicin aos gradientes "sen signo" que s toman valores entre 0 e 180 graos. No artigo orixinal do HOG utilizronse gradientes sen signo para a deteccin de pens.


## <font style="color:rgb(50,120,229)">Adestramento dun clasificador baseado en HOG para detectar gafas</font>
Nesta seccin, compartimos o cdigo a travs dos pasos para construr un clasificador baseado en HOG. O clasificador mira un vent na imaxe ao redor dos ollos e clasifcao como persoa con gafas ou sen elas.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0,10.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
predictions2Label = {0: "Sen gafas", 1: "Con gafas"}
```

## <font style="color:rgb(50,120,229)">Deteccin da cara e determinacin da rexin dos ollos</font>

Esta funcin emprega un Haar cascade xa entrenado en detectar caras frontais na imaxe. Unha vez atopada a rexin que contn a cara, determinamos a rexin dos ollo (heuristicamente sabemos que a altura superior e a inferior estn a 1/6 e 3/6 da altura da rexin da cara, respectivamente)

```python
#cargamos o haar-cascade entrenado para detectar caras frontais
face_cascade = cv2.CascadeClassifier('../data/models/haarcascade_frontalface_default.xml')

def getCroppedEyeRegion(targetImage):
    
    targetImageGray = cv2.cvtColor(targetImage, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(targetImageGray,1.3,5)
    x,y,w,h = faces[0]

    face_roi = targetImage[y:y+h,x:x+w]
    face_height,face_width = face_roi.shape[:2]

    # Aplicamos a formula heuristica para atopar
    # a rexion dos ollos unha vez atopada a rexion da cara
    eyeTop = int(1/6.0*face_height)
    eyeBottom = int(3/6.0*face_height)
    print("Rexion dos ollos entre comprendida entre: {},{}".format(eyeTop,eyeBottom))
    
    eye_roi = face_roi[eyeTop:eyeBottom,:]
    
    # Redimensionamos o tamanho dos ollos ao fixado na base de datos de 96x32
    cropped = cv2.resize(eye_roi,(96, 32), interpolation = cv2.INTER_CUBIC)
    
    return cropped
```

## <font style="color:rgb(50,120,229)">Preparacin dos datos</font>
A seguinte funcin le as imaxes nun directorio especificado. Crea das listas para almacenar imaxes de adestramento e test. Tamn crea listas para etiquetas de adestramento e test. O nmero de imaxes de proba (test) mantense nun 20% do nmero total de imaxes.

```python
# Path1 as clases 0 (sen gafas) e Path2 a clase 1(con gafas)
path1 = '../data/images/glassesDataset/cropped_withoutGlasses2'
path2 = '../data/images/glassesDataset/cropped_withGlasses2'

```

```python
def getTrainTest(path, class_val, test_fraction = 0.2):
  testData = []
  trainData = []
  trainLabels = []
  testLabels = []
  inputDir = os.path.expanduser(path)

  # Miramos cantas imaxes hai no directorio e atopamos o numero de adestramento e test
  if os.path.isdir(inputDir):
    images = os.listdir(inputDir)
    images.sort()
    nTest = int(len(images) * test_fraction)

  for counter, img in enumerate(images):

    im = cv2.imread(os.path.join(inputDir, img))
    # Engademos nTest mostras ao conxunto de test
    if counter < nTest:              
      testData.append(im)
      testLabels.append(class_val)
    else:   
      # Engadimos nTrain mostras ao conxunto de entrenamento
      trainData.append(im)
      trainLabels.append(class_val)

  return trainData, trainLabels, testData, testLabels
```

## <font style="color:rgb(50,120,229)">Configuracin do clasificador SVM</font>

O clasificador SVM inciase mediante a funcin **svmInit**. Establecemos os parmetros como:

- Tipo de ncleo
- Gama
- C

```python
def svmInit(C, gamma):
  model = cv2.ml.SVM_create()
  model.setGamma(gamma)
  model.setC(C)
  model.setKernel(cv2.ml.SVM_RBF)
  model.setType(cv2.ml.SVM_C_SVC)

  return model
```

O clasificador SVM adestrase mediante a funcin svmTrain. Os datos para a formacin deben proporcionarse no formato especfico que se indica a continuacin. Tamn gardamos o modelo aprendido nun ficheiro yml.

```python
def svmTrain(model, samples, responses):
  model.train(samples, cv2.ml.ROW_SAMPLE, responses)
  return model
```

Esta funcin sase para obter a sada ou predicin do clasificador.

```python
def svmPredict(model, samples):
  return model.predict(samples)[1].ravel()
```

Esta funcin atopa as predicins para todas as imaxes de proba da base de datos e atopa a precisin.

```python
def svmEvaluate(model, samples, labels):
  predictions = svmPredict(model, samples)
  accuracy = (labels == predictions).mean()
  print('Porcentaxe precision: %.2f %%' % (accuracy * 100))
  return accuracy
```

A seguinte funcin converte os descritores HoG de matriz a tipo Mat. Isto  necesario para o formato esixido por SVM

```python
def prepareData(data):
  featureVectorLength = len(data[0])
  features = np.float32(data).reshape(-1, featureVectorLength)
  return features
```

## <font style="color:rgb(50,120,229)">Configura o descritor HOG</font>

A seguinte funcin calcula os descritores HoG para unha determinada array de imaxes. Debemos especificar as imaxes nunha lista.

```python
def computeHOG(hog, data):

  hogData = []
  for image in data:
    hogFeatures = hog.compute(image)
    hogData.append(hogFeatures)

  return hogData
```

O conxunto de datos para o adestramento grdase no lugar especificado. As imaxes crganse mediante as funcins descritas anteriormente

Aqu, especificamos os parmetros do descritor HOG.

```python
# Inicializamos os parametros HoG
winSize = (96, 32)
blockSize = (8, 8)
blockStride = (8, 8)
cellSize = (4, 4)
nbins = 9
derivAperture = 0
winSigma = 4.0
histogramNormType = 1
L2HysThreshold = 2.0000000000000001e-01
gammaCorrection = 1
nlevels = 64

hog = cv2.HOGDescriptor(winSize, blockSize, blockStride, cellSize, 
                      nbins,derivAperture, winSigma, 
                      histogramNormType,L2HysThreshold, 
                      gammaCorrection, nlevels, 1)
```

## <font style="color:rgb(50,120,229)">Cargamos os datos nas variables de entrenamento e test</font>

```python
# Collemos as imaxes de entrenamento e test para as sas clases
negTrainImages, negTrainLabels, negTestImages, negTestLabels = \
getTrainTest(path1, 0, .2)
posTrainImages, posTrainLabels, posTestImages, posTestLabels = \
getTrainTest(path2, 1, .2)

# Engade imaxes positivas e negativas para adestrar e test
trainImages = np.concatenate((np.array(negTrainImages), 
                        np.array(posTrainImages)), 
                            axis=0)
testImages = np.concatenate((np.array(negTestImages), 
                          np.array(posTestImages)), 
                          axis=0)

# AEngade etiquetas positivas e negativas para adestrar e test
trainLabels = np.concatenate((np.array(negTrainLabels), 
                          np.array(posTrainLabels)), 
                          axis=0)
testLabels = np.concatenate((np.array(negTestLabels), 
                          np.array(posTestLabels)), 
                          axis=0)
```

## <font style="color:rgb(50,120,229)">Achamos as caracteristicas</font>

Aqu, calculamos os descritores HOG para os datos de adestramento e proba. A continuacin, convertemos os datos a un formato necesario para o clasificador SVM


```python
### Clculo de caractersticas para os datos de adestramento e proba  ##
trainHOG = computeHOG(hog, trainImages)
testHOG = computeHOG(hog, testImages)

# Converte os datos HoG en funcins recoecidas polo modelo SVM
trainFeatures = prepareData(trainHOG)
testFeatures = prepareData(testHOG)
```

## <font style="color:rgb(50,120,229)">Entrenamos o modelo SVM </font>

Aqu, especificamos os parmetros para SVM Classifier e adestramos o modelo usando as caracteristicas obtidas anteriormente.

```python
###########  adestramento SVM  ##############
model = svmInit(C=2.5, gamma=0.02)  # C = 0.1, gamma 10 for 
#collemos un kernel lineal e salvamos o clasificador entrenado
model = svmTrain(model, trainFeatures, trainLabels)
model.save("eyeGlassClassifierModel.yml")
```

Test e achamos a precisin do modelo para as imaxes de test

## <font style="color:rgb(50,120,229)">Avaliar o modelo e achamos a sa precisin</font>

```python
##########  SVM Test  ###############
# Cargaremos o modelo de novo e probaremos o modelo
# Isto  s para explicar como cargar un modelo SVM
# Podes usar o modelo directamente
savedModel = cv2.ml.SVM_load("eyeGlassClassifierModel.yml")

# Atopa a precisin do modelo
# avaliandoo sobre os datos da test
accuracy = svmEvaluate(savedModel, testFeatures, testLabels)
```

## <font style="color:rgb(50,120,229)">Comprobamos visualmente os resultados</font>

Aqu, usamos o modelo adestrado para realizar a clasificacin/predicin dunha nova imaxe. As funcins anteriores requiren que as imaxes estean nun vector. En primeiro lugar, recortamos a rexin dos ollos da imaxe usando os puntos de referencia dlib. A continuacin, convertemos a imaxe recortada nunha matriz e despois calculamos os descritores HOG. A continuacin, cargamos o modelo gardado e usamos a funcin **predict** para obter as predicins.

```python
# Realizamos un test sobre unha imaxe separada
filename = "../data/images/glassesDataset/glasses_4.jpg"
testImage = cv2.imread(filename)
cropped = getCroppedEyeRegion(testImage)
testHOG = computeHOG(hog, np.array([cropped]))
testFeatures = prepareData(testHOG)
predictions = svmPredict(savedModel, testFeatures)
print("Predicion = {}".format(predictions2Label[int(predictions[0])]))
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(testImage[:,:,::-1])
plt.subplot(1,2,2)
plt.imshow(cropped[:,:,::-1])
plt.show()
```
