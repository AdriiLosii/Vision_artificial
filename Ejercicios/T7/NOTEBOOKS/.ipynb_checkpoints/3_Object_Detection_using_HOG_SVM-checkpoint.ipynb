# <font style="color:rgb(50,120,229)">Detección de obxectos</font> 

No corazón de cada detector de obxectos hai un clasificador. A idea central do detector de obxectos é simplemente aplicar o clasificador a diferentes parches/ventás da imaxe.

A imaxe de abaixo é unha saída típica dun sistema de detección de peóns/persoas.

<center><img src="visuais/Det_obx.png" width=500/></center>

<center>Figure 1: Resultado típico dun detector de persoas/peons.</center> 


&nbsp;

A diferenza dun clasificador que actúa sobre toda a imaxe, na detección de obxectos necesitamos atopar a localización do obxecto na imaxe. Aquí van algunhas cousas que debes ter en conta cando deseñes un detector de obxectos.


1. **Localización:** o obxecto que estamos tentando detectar pode aparecer en calquera lugar dunha imaxe.

2. **Escala:** A escala ou o tamaño do obxecto pode ser arbitrario. No artigo de Dalal e Triggs, o detector de peóns foi adestrado para un parche de imaxe de tamaño 64 x 128. Non obstante, un peón pode aparecer nunha imaxe en calquera tamaño. Na imaxe superior, os dous xogadores detectados teñen unha escala moi diferente. Tamén temos que escoller o tamaño máis pequeno do obxecto que nos interesa apreciar. Na imaxe de arriba, probablemente non esteamos interesados en corredores que aparecen ao fondo da imaxe (moi pequenos).

3. **Rotación:** O obxecto pode non estar en posición vertical. Pode ser necesario rotar.

4. **Supresión non máxima:** Finalmente, a detección de obxectos require moitas veces un paso posterior ao procesamento chamado **supresión non máxima**. Se se detecta un obxecto nunha localización de píxel en particular, existe a posibilidade de que se detecte de novo nun píxel próximo. Estas dúas caixas delimitadoras apuntan ao mesmo obxecto e, polo tanto, simplemente necesitamos rexeitar as caixas delimitadoras superpostas. Isto chámase supresión non máxima.



## <font style="color:rgb(50,120,229)">Busqueda no espazo de escalas</font>

Detectar obxectos a varias escalas equivale a cambiar o tamaño da imaxe e utilizar o detector de tamaño fixo na imaxe redimensionada. Así, os detectores de obxectos crean internamente unha pirámide de imaxes que precisa de dous parámetros.

1. Niveis: o número de niveis da pirámide. Un número típico é 64.

2. Escala: Canto se redimensiona a imaxe entre os niveis da pirámide. Un número típico é 1,05.

Na maioría das implementacións dun detector de obxectos, a imaxe é reducida e nunca aumentada. Se se espera que o obxecto da túa imaxe sexa pequeno, debes ampliar a imaxe antes de enviala ao detector de obxectos e redimensionar o tamaño dos rectángulos contedores obtidos.

## <font style="color:rgb(50,120,229)">Búsqueda da posición</font>

A continuación, realizamos unha **busca mediante unha xanela deslizante** para a localización do obxecto en cada imaxe da pirámide. Suponse que o obxecto ten o mesmo tamaño que se empregou no adestramento. Por exemplo, mentres buscamos peóns, buscamos peóns de tamaño 64 x 128 porque non importa o tamaño real do peón, porque polo menos unha imaxe na pirámide conterá o peón próximo ao tamaño 64 x 128.

## <font style="color:rgb(50,120,229)">Clasificando unha ventá/parche</font>

Desde a subsección anterior, sabemos que varios parches da imaxe en varios lugares e a escala avalíanse para comprobar se hai algún obxecto dentro del.

A avaliación realízase mediante un clasificador que é capaz de decidir se un parche é un obxecto ou é fondo. Cando usamos un SVM como clasificador, as dúas clases están etiquetadas como -1 (fondo) e 1 (obxecto). Cando a resposta para o clasificador SVM é maior que 0, o parche ten unha maior probabilidade de pertencer á clase de obxecto. Na práctica, se a resposta do clasificador SVM é maior que un limiar (chamado **hitTreshold** en OpenCV) dicimos que é un obxecto. Un **hitTreshold** alto daría lugar a menos obxectos detectados.



## <font style="color:rgb(50,120,229)">Agrupación de caixas contedoras ( supresión de non-maximum )</font>

Como se mencionou antes, o detector pode detectar varios rectángulos para o mesmo obxecto. A agrupación destes rectángulos refírese a **supresión non máxima**.

Sabemos pola sección anterior que cada caixa delimitadora ten unha resposta SVM asociada. Estas respostas adoitan chamarse **pesos**. Canto maiores sexan os pesos, máis probable é que a caixa conteña un obxecto.

O obxectivo da supresión non máxima é crear unha lista pequena de rectángulos e o peso asociado a cada rectángulo da lista final.

Hai moitas formas diferentes de facer unha supresión non máxima. En OpenCV, o método **HOG::groupRectangles** emprega internamente os seguintes pasos.


   1. **Clustering**: Os rectángulos contedores agrúpanse en función da superposición. A idea é representar un clúster enteiro cun só rectángulo.

   2. **Pruning:** Se o número de rectángulos nun clúster é inferior a un limiar, elimínase o clúster.
    
   3. **Cluster average:** Atópase a media de todos os rectángulos de cada grupo. Este rectángulo engádese á lista final.
    
   4. **Collemos a máxima resposta do SVM:** A resposta do SVM (peso) asociada a este rectángulo medio é o máximo dos pesos de todos os rectángulos do clúster.
     
   5. **Filtrado de rectángulos pequenos:** A lista final aínda se acurta máis eliminando rectángulos máis pequenos contidos dentro doutros máis grandes se o número de rectángulos no cluster orixinal é inferior a 3.


<font style="color:rgb(255,0,0)">**Nota:**</font>A partir da descrición anterior, podemos ver que non todos os rectángulos completamente contidos dentro doutro rectángulo se eliminan pola supresión non máxima de OpenCV.

## <font style="color:rgb(50,120,229)">Como adestrar un detector de peóns/persoas baseado en HOG + SVM</font>

Nesta sección aprenderemos a adestrar un detector de obxectos baseado en HOG + SVM. Adestraremos un detector de persoas utilizando o conxunto de datos de peóns do INRIA. O cartafol contén dous subcartafoles:
1. train_64x128_H96: contén as imaxes recortadas dos peóns e o tamaño redúcese a 64x128 (Exemplos positivos)
2. test_64x128_H96: contén imaxes recortadas que non conteñen peóns. (Exemplos negativos)

Os datos de adestramento poden ser descargados dende **[esta ligazón](https://www.filepicker.io/api/file/VQKdmzKqSLWruVhx7Xdd)**. Unzip o ficheiro INRIAPerson.zip e sitúao en `../data/images/` antes de executar este código.

## <font style="color:rgb(50,120,229)">Utilidades de manexo de datos</font>

Funcións para ler imaxes dun directorio determinado e devolver as súas correspondentes etiquetas (+1 ou -1)

````python
# devolve os Paths das imaxes no cartafol indicado
# con extensións como se definen en imgExts
def getImagePaths(folder, imgExts):
  imagePaths = []
  for x in os.listdir(folder):
    xPath = os.path.join(folder, x)
    if os.path.splitext(xPath)[1] in imgExts:
      imagePaths.append(xPath)
  return imagePaths


# Le as imaxes dun cartafol
# devolve a lista de imaxes e etiquetas
def getDataset(folder, classLabel):
  images = []
  labels = []
  imagePaths = getImagePaths(folder, ['.jpg', '.png', '.jpeg'])
  for imagePath in imagePaths:
    # print(imagePath)
    im = cv2.imread(imagePath, cv2.IMREAD_COLOR)
    images.append(im)
    labels.append(classLabel)
  return images, labels
````

## <font style="color:rgb(50,120,229)">Configuramos o clasificador SVM</font>

Wrappers around SVM functions to initialize SVM, train on data and labels, predict labels and evaluate model’s error rate on test data

Inicializamos os parámetro de SVM, datos de adestramento e etiquetas, etiquetas a prdecir e como avaliar a taxa de erro do modelo a partir dos datos de test.

````python
# Inicializamos os parametros de SVM
def svmInit(C, gamma):
  model = cv2.ml.SVM_create()
  model.setGamma(gamma)
  model.setC(C)
  model.setKernel(cv2.ml.SVM_LINEAR)
  model.setType(cv2.ml.SVM_C_SVC)
  model.setTermCriteria((cv2.TERM_CRITERIA_EPS + 
                         cv2.TERM_CRITERIA_MAX_ITER, 
                         1000, 1e-3))
  return model


# Adestramos o SVM sobre os datos e as etiquetas
def svmTrain(model, samples, labels):
  model.train(samples, cv2.ml.ROW_SAMPLE, labels)


# predecimos as etiquetas sobre a mostra 
def svmPredict(model, samples):
  return model.predict(samples)[1]


# avaliamos o modelos comparando
# as predicións do modelo coas etiquetas do humano (ground truth)
def svmEvaluate(model, samples, labels):
  labels = labels[:, np.newaxis]
  pred = model.predict(samples)[1]
  correct = np.sum((labels == pred))
  err = (labels != pred).mean()
  print('label -- 1:{}, -1:{}'.format(np.sum(pred == 1), 
          np.sum(pred == -1)))
  return correct, err * 100


# Creamos un directorio se non existe
def createDir(folder):
  try:
    os.makedirs(folder)
  except OSError:
    print('{}: Xa existe'.format(folder))
  except Exception as e:
    print(e)
````

## <font style="color:rgb(50,120,229)">Inicializamos o detector de características HoG</font>

Funcións para calcular descritores HOG para un conxunto de imaxes e converter o descritor HOG ao formato de datos usado por SVM

````python
# achamos as características HOG para as imaxe
def computeHOG(hog, images):
  hogFeatures = []
  for image in images:
    hogFeature = hog.compute(image)
    hogFeatures.append(hogFeature)
  return hogFeatures


# convertemos os desctiptores HOG ao formato recoñecido por SVM
def prepareData(hogFeatures):
  featureVectorLength = len(hogFeatures[0])
  data = np.float32(hogFeatures).reshape(-1, featureVectorLength)
  return data
````

Initializamos os parámetros de HOG 


````python
# Parametros de HOG para inicializar
winSize = (64, 128)
blockSize = (16, 16)
blockStride = (8, 8)
cellSize = (8, 8)
nbins = 9
derivAperture = 1
winSigma = -1
histogramNormType = 0
L2HysThreshold = 0.2
gammaCorrection = True
nlevels = 64
signedGradient = False

# Inicializamos HOG
hog = cv2.HOGDescriptor(winSize, blockSize, blockStride,
                      cellSize, nbins,derivAperture,
                      winSigma, histogramNormType, L2HysThreshold, 
                      gammaCorrection, nlevels,signedGradient)
````

## <font style="color:rgb(50,120,229)">Inicializando os modos de entrenamento e test </font>

Flag para seleccionar o modo adestramento ou test ou query



````python
# Flasg para selecionar se entrenamos ou facemos test
trainModel = True
testModel = True
queryModel = True
````

Estamos usando o conxunto de datos de peóns do INRIA para adestrar o noso modelo. INRIA vén con cartafoles de entrenamento e test. Cada un destes cartafoles de train/test ten cartafoles pos e neg que conteñen imaxes positivas e negativas. Dado que SVM está adestrado usando vectores do mesmo tamaño, creamos parches do mesmo tamaño desde os directorios pos e neg ata posPatches e negPatches. negPatches conteñen 10 parches seleccionados aleatoriamente de cada imaxe do directorio neg. posPatches contén 1 parche para cada imaxe no directorio pos despois de eliminar o recheo (16 para o entrenamento e 3 para test).

````python
# Path ao dataset de persoas do INRIA
rootDir = '../data/images/INRIAPerson/'

# Asignamos os directorio de entrenamento e test
trainDir = os.path.join(rootDir, 'train_64x128_H96')
testDir = os.path.join(rootDir, 'test_64x128_H96')
````

## <font style="color:rgb(50,120,229)">Adestramento</font>

Le imaxes dos directorios posPatches e negPatches e crea os arrays de etiquetas correspondentes. +1 para pos e -1 para neg. Finalmente fusionamos vectores de imaxes e etiquetas mantendo as mostras negativas despois das mostras positivas.

````python
# ================================ Entrenamento do modelo =====================
if trainModel:
    # Lemos as imaxes dos directorio de Pos e Neg 
    trainPosDir = os.path.join(trainDir, 'posPatches')
    trainNegDir = os.path.join(trainDir, 'negPatches')

    # Label 1 fpara imaxes positivas e -1 as negativas
    trainPosImages, trainPosLabels = getDataset(trainPosDir, 1)
    trainNegImages, trainNegLabels = getDataset(trainNegDir, -1)

    # Comprobamos os tamaños das imaxes so iguais
    print(set([x.shape for x in trainPosImages]))
    print(set([x.shape for x in trainNegImages]))

    # visualizamos o numero de imaxes positivas e negativas que lemos
    print('positivas - {}, {} || negativas - {}, {}'
        .format(len(trainPosImages),len(trainPosLabels),
        len(trainNegImages),len(trainNegLabels)))

    # Engadimos Positive/Negative Images/Labels para adestramento
    trainImages = np.concatenate((np.array(trainPosImages), 
                       np.array(trainNegImages)), 
                                  axis=0)
    trainLabels = np.concatenate((np.array(trainPosLabels), 
                       np.array(trainNegLabels)),
                                  axis=0)
    # Achamos as caracteristicas HOG para adestramento e 
    # convertemos o descriptor HOG ao formato esperado por SVM.
    # Acahamos HOG para as imaxes
    hogTrain = computeHOG(hog, trainImages)

    # Formato para SVM
    trainData = prepareData(hogTrain)

    # Comprobamo as dimensions dos datoa e as etiquetas
    print('trainData: {}, trainLabels:{}'
            .format(trainData.shape, trainLabels.shape))
    # Creamos o obxecto SVM, entrenamos o modelos e gardamolo.
    model = svmInit(C=0.01, gamma=0)
    svmTrain(model, trainData, trainLabels)
    model.save('pedestrian.yml')
````

## <font style="color:rgb(50,120,229)">Test</font>
Cargamos o modelo entrenado dende disco 

````python
# ================================ Test o modelo ===============
if testModel:
    # cargamos o modelo dende disco
    model = cv2.ml.SVM_load('pedestrian.yml')
    # Avaliaremos o noso conxunto de datos de proba para positivod e
    # imaxes negativas por separado para calcular verdadeiros/falsos positivos
    # e verdadeiros/falsos negativos
    testPosDir = os.path.join(testDir, 'posPatches')
    testNegDir = os.path.join(testDir, 'negPatches')
    
    # Seguiremos o mesmo fluxo para preparar datos que fixemos para as imaxes de adestramento.
     # Etiqueta 1 para imaxes positivas e -1 para imaxes negativas
    testPosImages, testPosLabels = getDataset(testPosDir, 1)
    testNegImages, testNegLabels = getDataset(testNegDir, -1)

    # HoG para as imaxes
    hogPosTest = computeHOG(hog, np.array(testPosImages))
    testPosData = prepareData(hogPosTest)
    
    # Executa a clasificación nos datos da test para as imaxes positivas
    # e calcula os verdadeiros positivos e os falsos positivos.
    posCorrect, posError = svmEvaluate(model, testPosData, 
                                       np.array(testPosLabels))

    # Calcula falsos positivos e negativos
    tp = posCorrect
    fp = len(testPosLabels) - posCorrect
    print('TP: {}, FP: {}, Total: {}, error: {}'
            .format(tp, fp, len(testPosLabels), posError))

    # Do mesmo xeito, executa a clasificación dos datos negativos (test)
    # e calcula verdadeiros positivos e os falsos positivos. 
    hogNegTest = computeHOG(hog, np.array(testNegImages))
    testNegData = prepareData(hogNegTest)
    negCorrect, negError = svmEvaluate(model, testNegData, 
                                       np.array(testNegLabels))

    # falsos negativos e positivos
    tn = negCorrect
    fn = len(testNegData) - negCorrect
    print('TN: {}, FN: {}, Total: {}, error: {}'
            .format(tn, fn, len(testNegLabels), negError))
    # Achamos Precision e Recall
    precision = tp * 100 / (tp + fp)
    recall = tp * 100 / (tp + fn)
    print('Precision: {}, Recall: {}'.format(precision, recall))
````

## <font style="color:rgb(50,120,229)">Test o modelo sobre unha nova imaxe </font> 
Queremos cargar os modelos SVM adestrados por nós e utilizalos para atopar peóns nunha nova imaxe. Para iso, necesitamos utilizar a función detectMultiScale que aporta OpenCV.

## <font style="color:rgb(50,120,229)">Uso de **hog.detectMultiScale**</font>

Unha vez cargado o clasificador no detector, a función **detectMultiScale** úsase para detectar obxectos nunha imaxe.

### <font style="color:rgb(8,133,37)">Sintaxe da función</font>

```python
(foundLocations, foundWeights) = hog.detectMultiScale(img, hitThreshold=hitThreshold, winStride=winStride, padding=padding, scale=scale,  finalThreshold=finalThreshold, useMeanshiftGrouping=useMeanshiftGrouping)
```

onde

- **`img`** = Imaxe de entrada

- **`foundLocations`** = Un vector cos rectángulos contedores aos correspondentes a obxectos detectados


- **`foundWeights`** = SVM resposta (peso) asociada ás caixas delimitadoras anteriores. Podes tratar este peso como unha medida de confianza.


- **`hitThreshold`** = O limiar de resposta SVM por encima do cal dicimos que o rectángulo é un obxecto. Normalmente, o valor predeterminado de 0 é bo. Observa que os elementos de foundWeights serán sempre maiores que hitThreshold.

- **`winStride`** = O paso da xanela de detección de obxectos como se explicaou anteriormente

- **`padding`** = Engádese un recheo a toda a imaxe para que se poidan detectar obxectos próximos ao límite. Normalmente establécese á metade do ancho da xanela de detección. p.ex. na detección de peóns onde a xanela de detección ten un tamaño de 64 x 128, o recheo está configurado en 32 x 32.

- **`scale`** = Para crear a pirámide de imaxes, a imaxe nun nivel é escalada por este número. p.ex. cando a escala se establece en 1,05, significa que o segundo nivel da pirámide se crea cambiando o tamaño da imaxe orixinal nun factor de 1/1,05. O terceiro nivel créase reescalando a imaxe no nivel 2 nun factor de 1/1,05 e así sucesivamente.


- **`finalThreshold`** = Un nome mellor para este parámetro sería groupThreshold. No paso de supresión non máxima, este parámetro utilízase para eliminar os clústeres que teñen menos número de rectángulo que os indicados por finalTreshol. 

- **`useMeanshiftGrouping`** = Use o algoritmo de agrupación non estándar baseado sobre mean-shift. En base á lectura do código e da documentación, non está claro cando é unha boa idea.

````python

# ================================= Consulta/Query ================ ================================
# Executar o detector de obxectos nunha imaxe de consulta para atopar peóns
# Cargaremos o modelo de novo e probaremos o modelo
# Isto é só para explicar como cargar un modelo SVM
# Podes usar o modelo directamente dende memoria (xa o temos aí)
if queryModel:
    # cargamos o modelos
    model = cv2.ml.SVM_load('../data/models/pedestrian.yml')
    # Extraemos o SVM e o bias (sesgo) dende modelo
    sv = model.getSupportVectors()
    rho, aplha, svidx = model.getDecisionFunction(0)
    svmDetector = np.zeros(sv.shape[1] + 1, dtype=sv.dtype)
    svmDetector[:-1] = -sv[:]
    svmDetector[-1] = rho

    # Iniciamos o noso SVMDetector en HOG
    hog.setSVMDetector(svmDetector)

    filename = "../data/images/pedestrians/3.jpg"
    queryImage = cv2.imread(filename, cv2.IMREAD_COLOR)

    #executamos o noso detector nunha imaxe cunha altura fixa
    finalHeight = 800.0
    scale = finalHeight / queryImage.shape[0]
    queryImage = cv2.resize(queryImage, None, fx=scale, fy=scale)


    # detectMultiScale detectará a n niveis de imaxe aumentando a escala
     # e reducindo o tamaño da imaxe por un factor de 1,05
    bboxes, weights = hog.detectMultiScale(queryImage, winStride=(8, 8),
                                           padding=(32, 32), scale=1.05,
                                           finalThreshold=2, hitThreshold=1.0)
    # Visualizamos as caixas na iamxe
    for bbox in bboxes:
      x1, y1, w, h = bbox
      x2, y2 = x1 + w, y1 + h
      cv2.rectangle(queryImage, (x1, y1), (x2, y2),
                    (0, 255, 0), thickness=3, lineType=cv2.LINE_AA)
    plt.imshow(queryImage[:,:,::-1])
    plt.show()
````

## <font style="color:rgb(50,120,229)">Comparación do noso modelo co dectector de persoa implementado en OpenCV</font>

Ata agora probamos o noso modelo en parches de imaxe de 64 x 128. Executaremos o noso modelo con algunhas imaxes da vida real. Tamén compararemos os nosos resultados co Detector de persoas predeterminado de OpenCV. O detector de persoas de OpenCV tamén está adestrado no conxunto de datos do INRIA. Debuxaremos rectángulos verdes para os resultados do noso modelo e caixas vermellas para os resultados do detector de persoas de OpenCV.

````python
# Iniciamos HOG
winSize = (64, 128)
blockSize = (16, 16)
blockStride = (8, 8)
cellSize = (8, 8)
nbins = 9
derivAperture = 1
winSigma = -1
histogramNormType = 0
L2HysThreshold = 0.2
gammaCorrection = True
nlevels = 64
signedGradient = False

hog = cv2.HOGDescriptor(winSize, blockSize, blockStride,
                        cellSize, nbins, derivAperture,
                        winSigma, histogramNormType, L2HysThreshold,
                        gammaCorrection, nlevels, signedGradient)
````


Cargamos o modelo, extraemos vectores de apoio e función de decisión (rho) deste modelo. Agora creamos un detector SVM para HOG engadindo rho ao final dos vectores de soporte. Agora configuramos este detector SVM en HOG.

````python
# Cargamos o modelo entrenao por nos
model = cv2.ml.SVM_load('pedestrian.yml')
sv = model.getSupportVectors()
rho, aplha, svidx = model.getDecisionFunction(0)
svmDetectorTrained = np.zeros(sv.shape[1] + 1, dtype=sv.dtype)
svmDetectorTrained[:-1] = -sv[:]
svmDetectorTrained[-1] = rho
# iniciamos SVMDetector entrenado por nos en HOG
hog.setSVMDetector(svmDetectorTrained)
````

Inicialice HOG e configura o detector de persoas predeterminado de OpenCV con SVM.


````python
# Detector de persoas de OpenCV con HoG
hogDefault = cv2.HOGDescriptor(winSize, blockSize, blockStride,
                               cellSize, nbins, derivAperture,
                               winSigma, histogramNormType, 
                               L2HysThreshold,gammaCorrection,
                               nlevels, signedGradient)
svmDetectorDefault = cv2.HOGDescriptor_getDefaultPeopleDetector()
hogDefault.setSVMDetector(svmDetectorDefault)
````

Le imaxes para o directorio de peóns que ten algunhas imaxes do mundo real. Realizaremos a detección en imaxes con altura = 800


````python
# lemos as imaxes do directorio de persoas
imageDir = '../data/images/pedestrians'
imagePaths = glob.glob(os.path.join(imageDir, '*.jpg'))

#executamos o detector para imaxes con altura fixa
finalHeight = 800.0

for imagePath in imagePaths:
    print('procesando: {}'.format(imagePath))

    # Lemos a imaxe
    im = cv2.imread(imagePath, cv2.IMREAD_COLOR)

    # redimensionamos a imaxe para que teña unha altura prederminada de finalHeight
    scale = finalHeight / im.shape[0]
    im = cv2.resize(im, None, fx=scale, fy=scale)
    
    # Detecta persoas usando detectores SVM adestrados e predeterminados
    # detectMultiScale usando un detector adestrado por nós
    bboxes, weights = hog.detectMultiScale(im, winStride=(8, 8),
                                    padding=(32, 32),scale=1.05, 
                                    finalThreshold=2,hitThreshold=1.0)

    # detectMultiScale modelos predetemiando
    bboxes2, weights2 = hogDefault.detectMultiScale(im, winStride=(8, 8), 
                                padding=(32, 32),scale=1.05, 
                                finalThreshold=2,hitThreshold=0)
    # Debuxa rectángulos atopados na imaxe. Debuxaremos
     # caixas verdes para persoas detectadas polo modelo adestrado e
     # caixas vermellas para as persoas detectadas polo modelo predeterminado de OpenCV.
    if len(bboxes) > 0:
        print('Detector entrenado :: peons detectados: {}'
                .format(bboxes.shape[0]))
    if len(bboxes2) > 0:
        print('Detector de OpCV :: peons detectados: {}'
                .format(bboxes2.shape[0]))


    # Debuxa caixas delimitadoras detectadas sobre a imaxe
    # Vermello = detector predeterminado, Verde = Detector adestrado por nos
    for bbox in bboxes:
        x1, y1, w, h = bbox
        x2, y2 = x1 + w, y1 + h
        cv2.rectangle(im, (x1, y1), (x2, y2), 
                      (0, 255, 0), thickness=3, 
                      lineType=cv2.LINE_AA)

    for bbox in bboxes2:
        x1, y1, w, h = bbox
        x2, y2 = x1 + w, y1 + h
        cv2.rectangle(im, (x1, y1), (x2, y2), 
                      (0, 0, 255), thickness=3, 
                      lineType=cv2.LINE_AA)

        # Visualizamos os resultados finais
    plt.imshow(im[:,:,::-1])
    # a disco ...
    imResultPath = os.path.join('results', os.path.basename(imagePath))
    cv2.imwrite(imResultPath, im)
    plt.show()
````


### <font style="color:rgb(50,120,229)">Minería de negativos difíciles</font>

Podes ter notado que o noso modelo produce un pouco máis de falsos positivos. É porque non fixemos minería negativa. Para a minería negativa só tes que executar este detector en imaxes negativas (Podes descargar algunhas imaxes que non conteñan persoas).

O modelo está dando falsos positivos porque é difícil predecir estes parches correctamente. Entón, queremos adestralo sobre algúns exemplos difíciles. Isto tamén se denomina "Mining Hard Negative".

Cando executa o detector en imaxes negativas completas, escribe todos aqueles parches ( bbox ) que se prevén como positivos (é dicir, etiqueta prevista = 1) nun directorio (chamémoslle falsePosPatches). Asegúrate de cambiar o tamaño deste parche a 64x128 antes de escribir no directorio falsePosPatches.

Agora adestra o modelo de novo esta vez con imaxes en posPatches como imaxes positivas e imaxes en negPatches + falsePosPatches como imaxes negativas.
