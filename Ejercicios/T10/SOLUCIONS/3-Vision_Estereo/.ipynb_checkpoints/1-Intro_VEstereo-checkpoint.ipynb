## 3.1.2 Triangulacin e disparidade <span id=312></span>

Comecemos describindo como obter as coordenadas 3D dun determinado punto a partir dos seus pxeles correspondentes nas imaxes. Para iso, utilizaremos os coecementos do caderno onde indicamos o modelo de cmara, e asumiremos **unha configuracin de cmara ideal**,  dicir:

- Cmaras idnticas e paralelas
- Cmara dereita ao longo do eixe $X_l$ a unha distancia $b$ (chamada *lnea de base*)

Non importa que, na prctica, as nosas cmaras non estean exactamente colocadas como se describe, porque sempre podemos **rectificar as imaxes** mediante a chamada **xeometra epipolar**, que se abordar no seguinte caderno. Pero, polo momento, supoamos que temos unha cmara estreo ideal.$\\[5pt]$

Neste caso, a relacin xeomtrica entre as das cmaras di o seguinte:$\\[5pt]$

$$X_l = \mathbf{R}X_r + \mathbf{t} \qquad \mathbf{R} = \mathbf{I} \qquad \mathbf{t} = \begin{bmatrix}b \\ 0 \\ 0 \end{bmatrix}$$

<img src="./images/triangulation1.png" />

Ten en conta que a pose da cmara esquerda indicada por $X_l$ est relacionada coa pose da cmara dereita $X_r$ a travs dunha matriz de rotacin e un vector de translacin, que, neste caso, son a **matriz de identidade** e un **translacin** nica no eixe $X$, respectivamente. Como tamn podedes ver, nesta configuracin os pxeles correspondentes a un determinado punto 3D teen en comn as sas coordenadas *filas*,  dicir, **proxctanse  mesma lia de pxeles en ambas imaxes: $y_r = y_l$** . Esta propiedade  extremadamente til cando se buscan correspondencias.

As, dado un punto 3D $\bf{P} = (X,Y,Z)$, podemos obter as coordenadas proxectadas tanto nas imaxes da esquerda como da dereita por triangulacin como:$\\[5pt]$


<img src="./images/triangulation2.png" width="300" align="left"/>$\\[10pt]$

$$\large{\begin{cases} \frac{x_l}{f} = \frac{X_l}{Z} \\[10pt]  \frac{x_r}{f} = \frac{X_r}{Z} = \frac{X_l - b}{Z} \end{cases}} \normalsize{\xrightarrow{\text{Eliminamos } X_l}  \frac{b}{Z} = \frac{\color{red}{x_l- x_r}}{f} \rightarrow Z = b \frac{f}{\color{red}{d}}} \quad \small{\text{(}d \text{ = disparidade)}}\\[20pt]$$

$$\text{Eixe Y (o mesmo para ambos eixes):} \qquad \frac{y_l}{f} = \frac{y_r}{f} = \frac{Y}{Z} \hspace{6cm}\\[10pt]$$

$$
\begin{align}
\mathbf{P} &= \left[X \quad Y \quad Z\right]^T 
= \left[ Z\frac{x_l}{f} \quad Z\frac{y_l}{f} \quad Z \right]^T  \\
&= \left[ b\frac{x_l}{d} \quad b\frac{y_l}{d} \quad b\frac{f}{d} \right]^T 
= \frac{b}{d}\left[x_l \quad y_l \quad f\right]^T 
\end{align}
\\[30pt]$$


Nesta ecuacin introducimos o concepto de **disparidade**, que representa a diferenza na coordenada $x$ dos puntos proxectados en ambas imaxes,  dicir, $d = x_l-x_r$. Ten en conta que a profundidade $\mathbf{Z}$  inversamente proporcional  disparidade $d$. Ten en conta tamn que, nesta ecuacin, **tanto a disparidade como as coordenadas $(x_l,y_l)$ estn en metros**, medidos no sensor. Lembra que, se queremos usar pxeles como unidades, temos que facer unha **transformacin do sensor da cmara  matriz de imaxe**:$\\[5pt]$
<img src="./images/sensor_matrix.png" width="600"/>$\\[5pt]$

$$\begin{eqnarray} 
    u & = & xk_x + u_o \rightarrow x & = & (u - u_0)/k_x \\ 
    v & = & yk_y + v_o \rightarrow y & = & (v - v_0)/k_y 
\end{eqnarray} \longrightarrow d = x_l - x_r = \frac{(u_l - u_r)}{k_x} = \frac{d'}{k_x} \qquad d' \ \text{significa disparidade en pixeles}\\[15pt]$$

Para que cheguemos a:

$$
\boxed{\mathbf{P} = \frac{b}{d}\left[x_l \quad y_l \quad f\right]^\texttt{T} 
             = \frac{bk_x}{d'}\left[\frac{(u_l - u_0)}{k_x}\quad \frac{(v_l - v_0)}{k_y}\quad f\right]^\texttt{T} 
            = \frac{b}{d'}\left[(u_l - u_0)\quad \frac{k_x}{k_y}(v_l - v_0)\quad k_xf\right]^\texttt{T}}\\[30pt]
$$

Como podes ver, os parmetros intrnsecos da cmara $(u_0,v_0,k_x,k_y,f)$, as como a *lnea de base* (que  un parmetro extrnseco), son necesarios para calcular as coordenadas 3D dunha correspondencia en dous imaxes. **Por iso necesitamos calibrar a cmara**.

### Precisin da reconstrucin

Agora sabemos como obter puntos 3D das sas correspondencias, podemos preguntarnos **que precisin  o punto 3D reconstrudo?** Ben, a principal fonte de erro provn das **imprecisins na deteccin dos puntos**, que provocan erros nas coordenadas 3D, xa que ao calcular a profundidade, tanto a distancia focal como a lia base teen valores fixos e a nica **variable**  a **disparidade**. Por suposto, os erros na determinacin da distancia focal ou da lia de base inflen no erro final, pero pdense estimar con mis precisin. $\\[5pt]$

$$Z = b\frac{f}{d} \rightarrow \text{erro na deteccion dos puntos} \rightarrow \text{erro na disparidade} \rightarrow \text{erro na profundidade}\\[5pt]$$


As, se temos un erro de disparidade $\Delta d$,  posible calcular como este erro se propaga a $Z$ calculando a derivada da coordenada con respecto  profundidade:$\\[5pt]$

$$Z = b\frac{f}{d} \rightarrow \frac{\Delta Z}{\Delta d} = -\frac{bf}{d^2} \longrightarrow \Delta Z = -\frac{bf}{d}\Delta d = - \frac{Z^2}{fb}\Delta d\\[5pt]$$

Polo tanto, dado un erro de disparidade $\Delta d$, o erro en profundidade **crece cuadrticamente** co valor $Z$. Isto implica que para obxectos mis prximos ($\downarrow Z$), a influencia do erro de disparidade vai ser menor.  dicir, a posicin dos puntos mis prximos  cmara estimarase con mis precisin que a dos puntos mis afastados.

Por outra banda, para aumentar a precisin dos puntos 3D, poderase aumentar a lia de base ( dicir, separando as cmaras), pero isto pode causar mis dificultades para atopar correspondencias xa que **as imaxes deben superpoerse!** De feito, os sistemas estreo usado para aplicacins ao aire libre adoita presentar lias de base mis amplas xa que estn destinados a utilizar puntos afastados, mentres que os pares estreo de lia base pequena resrvanse para aplicacins en interiores. $\\[5pt]$

## 3.1.3 Restricins para a correspondencia <span id=313></span>

A primeira e unha das decisins mis importantes antes de comezar a nosa tarefa  decidir **que caractersticas imos detectar e coincidir**. Podemos combinar todos os pxeles das imaxes (*estreo denso*), s os puntos clave (*estreo dispersin*) ou outras caractersticas da imaxe como bordes, segmentos, rexins, etc. E tamn temos que decidir **que descritor** utilizaremos (Harris, SIFT/ORB/SURF para puntos clave, etc).

Un dos criterios mis importantes  hora de elixir isto  **robusteza de coincidencia**,  dicir, deben ser o suficientemente distintivos como para evitar calquera coincidencia incorrecta, xa que isto pode resultar catastrfico para a nosa aplicacin. Normalmente, estas coincidencias incorrectas chmanse **outliers** e deben evitarse na medida do posible, se non, xeraremos puntos 3D *inventados*:


<img src="./images/correspondence1.png" width="500"/>$\\[5pt]$

Como podes ver na imaxe, se coincidimos o punto $d$ da imaxe da esquerda co punto $a'$ da dereita **xeraremos un punto 3D na escena que non pertence a ningn obxecto dela**. Se mis tarde empregamos ese punto 3D para localizar a nosa cmara, por exemplo, obteremos unha estimacin errnea de tal localizacin.

Para filtrar os valores atpicos, podemos aplicar algunhas restricins do entorno:

### Permtese a disparidade mxima e mnima

Esta  unha limitacin importante, xa que podes establecer facilmente a **disparidade mxima e mnima** para unha determinada rea da escena, dependendo da profundidade dos obxectos nela, as como da lia base e da distancia focal do noso par estreo. Deste xeito, se unha coincidencia **presenta unha disparidade fra dos nosos lmites preestablecidos ser descartada**.

<img src="./images/correspondence2.png" width="500"/>$\\[5pt]$


### Continuidade das superficies

No mundo real, **as superficies son contnuas**,  dicir, a profundidade cambia suavemente. Isto s se viola nos bordes de oclusin, que se poden detectar usando un detector de bordes como Canny. Pero ademais dos bordes, haber unha alta probabilidade de que **os pxeles prximos tean unha profundidade (e, polo tanto, disparidade) semellantes**.

<img src="./images/correspondence3.png" width="400"/>$\\[5pt]$

### Singularidade

Cada pxel da imaxe  a proxeccin de **s un punto en 3D**. Noutras palabras, un pxel dunha imaxe s ter un pxel correspondente na outra imaxe. Esta  a razn pola que, se lembras, descartamos as coincidencias ambiguas con SIFT establecendo un determinado limiar para que a mellor correspondencia debe estar o suficientemente afastada do resto de posibilidades.

<img src="./images/correspondence4.png" width="400"/>$\\[5pt]$

### Ordenacin e xeometra epipolar

Abordaremos estas restricins en futuros cadernos, pero a grandes lias, o primeiro indica que os puntos correspondentes ao longo dunha lia (chamada lia epipolar) **deben atoparse na mesma orde en ambas as imaxes** (con algunhas excepcins).

<img src="./images/correspondence5.png" width="300"/>$\\[5pt]$

A segunda est baseada na chamada xeometra epipolar e indica que o punto correspondente para un determinado pxel nunha imaxe **debe estar nunha soa lia** (a lia epipolar de novo) na outra imaxe, reducindo as efectivamente o rea de busca de 2D a 1D.

### **<span style="color:green"><b><i>TAREFA 1: Tomando e cargando imaxes para facer estreo</i></b></span>**

En primeiro lugar, imos **tomas 2 imaxes** (esquerda e dereita) de calquera escena (por exemplo, colocar algns obxectos nunha mesa). Para simular as condicins necesarias para a triangulacin, intenta mover a cmara hacia  dereita pero **sen xirarla** (recorda que nunha configuracin ideal, a matriz de rotacin entre as cmaras  a identidade e a traslacin s  un desfase na direccin $X$). Mide e anota a distancia que moviches a cmara entre as das imaxes (esta  a nosa *lnea de base*).

<img src="./images/ideal_conf_example.png" width="300"/>$\\[5pt]$

Para ser precisos, debemos **rectificar** as imaxes, pero ese proceso explicarase no seguinte caderno. De momento, intentemos facer estreo sen rectificar as imaxes, pero tendo en conta que **os resultados non sern moi precisos**, porque non se eliminar a distorsin e a configuracin espacial das vistas non ser a perfecta. Isto eventualmente implica que os **pxeles correspondentes non estarn na mesma coordenada horizontal**, senn preto dunha.

**Agora, carga as imaxes da esquerda e da dereita**.

*Nota: Lembra que podes usar os exemplos.*

```python
# Tarefa 1a
# Cargamos as imaxes
# Imaxes do exemplo:
# - iml_sword.png, imr_sword.png
# - iml_lamp.png, imr_lamp.png

im_l = cv2.imread(images_path + 'iml_sword.png')
im_l = cv2.cvtColor(im_l, cv2.COLOR_BGR2RGB)

im_r = cv2.imread(images_path + 'imr_sword.png')
im_r = cv2.cvtColor(im_r, cv2.COLOR_BGR2RGB)

# Imaxe esquerda
plt.subplot(121)
plt.imshow(im_l)
plt.title('Imaxe esquerda')

# Imaxe dereita
plt.subplot(122)
plt.imshow(im_r)
plt.title('Imaxe dereita');
```

### **<span style="color:green"><b><i>TAREFA 2: Deteccin e coincidencia de puntos clave</i></b></span>**

Unha vez que teamos as nosas imaxes, decidimos utilizar **puntos clave** para realizar estereo. Deste xeito, a reconstrucin 3D s mostrar os puntos clave coincidentes,  dicir, ser *dispersa*. **A ta tarefa** aqu  usar o teu **detector de puntos clave + descritor** favoritos (escolle un do captulo 6) para obter un **conxunto de puntos coincidentes** que se utilizarn para a triangulacin.

**Utiliza o mtodo escollido para detectar e facer coincidir os puntos en ambas imaxes.**

*Nota: para imaxes grandes,  ORB  unha boa opcin pola sa rapidez.*

```python
# TAREFA 2
# Detecta puntos clave e busca as coincidencias na outra imaxe

# cargamos as imaxes en gris
gray_l = cv2.cvtColor(im_l,cv2.COLOR_RGB2GRAY)
gray_r = cv2.cvtColor(im_r,cv2.COLOR_RGB2GRAY)

# Detecta os puntos clave ORB usando o mtodo OpenCV
# -- crear o detector ORB
orb = cv2.ORB_create()

# -- detectar puntos clave ORB
kp_l = orb.detect(gray_l,None) # Deixa o segundo none como est
kp_r = orb.detect(gray_r,None) # Deixa o segundo none como est

# -- calcular os descriptores con ORB
kp_l, des_l = orb.compute(im_l, kp_l)
kp_r, des_r = orb.compute(im_r, kp_r)

# Nota: a deteccin e a descricin pdense facer nunha soa chamada

# Facemos o match
matches = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True).match(des_l,des_r)

# Ordnaos segundo a sa distancia.
matches = sorted(matches, key = lambda x:x.distance)
output = np.zeros_like(gray_r)
# Representa os matches nas duas imaxes
imMatches = cv2.drawMatches(gray_r, kp_r, gray_l, kp_l, matches, output)
plt.figure()
plt.title("Matches")
plt.imshow(imMatches);
```

### **<span style="color:green"><b><i>TAREFA 3: Filtrando coincidencias e obtencin de coordenadas 3D</i></b></span>**

Para iso, estamos a usar a restricin dos **puntos correspondentes situados na mesma *fila* coordenada** (cun certo limiar, lembra que a configuracin non  exactamente a ideal!). Usa un limiar (por exemplo, $|v_l-v_r| < th$) para manter s as coincidencias que cumpren a restricin. Tamn podes definir un **rango de disparidade** a ter en conta.

Unha vez que filtras as tas coincidencias, obtn as sas coordenadas 3D aplicando a **triangulacin** como se explica anteriormente. Para iso, necesitar os parmetros intrnsecos e extrnsecos do par estreo ou cmara.

**Que facer?** Mostra unha das imaxes e debuxa nela as coincidencias filtradas (podes usar [cv2.circle](https://docs.opencv.org/4.2.0/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670)). Ademais, realiza a triangulacin para obter unha lista das coordenadas 3D de tales puntos.

**Deberas esperar unha sada como esta:**

<img src="./images/matched_keypoints.png"/>

```python
# TAREFA 3
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15.0, 15.0)

# Asumimos que kx/ky=1

# base de datos SWORD
baseline = 17.4724 # cm
intrinsic_l = np.array([[6872.874, 0, 1329.49],[0, 6872.874, 954.485],[0, 0, 1]]) # pixels
intrinsic_r = np.array([[6872.874, 0, 1623.46],[0, 6872.874, 954.485],[0, 0, 1]]) # pixels


d_min = 0
d_max = 20000

# Define auxiliar data structures
points3D = []
color3D  = []

im_points_l = im_l.copy() # imaxe esquerda para a sada
im_points_r = im_r.copy() # imaxe da dereita para a sada

f  = intrinsic_l[0,0]
u0 = intrinsic_l[0,2]
v0 = intrinsic_l[1,2]

# para cada match
for match in matches:
    point_l = kp_l[match.queryIdx].pt # [0] --> x (col), [1] -- > y (row)
    point_r = kp_r[match.trainIdx].pt
    
    # comprobar as coordenadas da fila
    if abs (point_l[1]-point_r[1]) < 10: 
        # print("Match:",point_l,"vs",point_r)
        # comprobar o filtro de disparidade
        d = point_l[0] - point_r[0] # disparidade
        if d < d_max and d > d_min:
            # Debuxa un crculo
            cv2.circle(im_points_l,(int(point_l[0]),int(point_l[1])),15,[255,0,0],cv2.FILLED)
            cv2.circle(im_points_r,(int(point_r[0]),int(point_r[1])),15,[255,0,0],cv2.FILLED)
            
            # Obtn a cor do pxel para 3D
            color3D.append(im_l[int(point_l[1]),int(point_l[0]),:])
            
            # Obter a proxeccin 3D mediante a triangulacin
            points3D.append(( baseline/d * (point_r[0]-baseline),
                              baseline/d * (point_r[1]-baseline),
                              baseline/d * f))

# Mostra cales son as coincidencias que cumpren a condicin
plt.subplot(121)
plt.title("Imaxe esquerda")
plt.imshow(im_points_l);

plt.subplot(122)
plt.title("Imaxe dereita")
plt.imshow(im_points_r);
```

### **<span style="color:green"><b><i>TAREFA 4: Mostrando a nosa escena reconstruda</i></b></span>**

Unha vez calculadas as coordenadas 3D, o ltimo paso  **mostrar unha grfica de dispersin en cores 3D da escena** utilizando os datos obtidos no paso anterior.

**Exemplo de reconstrucin 3D:**
    

<img src="./images/3d_reconstruction.png"/>

```python
# TAREFA
# %matplotlib notebook
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)

# Transformar datos en matrices NumPy
points3D = np.array(points3D, dtype=np.int32)
color3D  = np.array(color3D, dtype=np.int32)

# Crear figura
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection='3d')

# Nomear eixes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Representa os puntos
ax.scatter(points3D[:,0], points3D[:,1], points3D[:,2], c=color3D/255, cmap="rgb")
ax.view_init(elev=0, azim=-90)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_zlabel('Z axis')

# Representa os puntos
ax2.scatter(points3D[:,0], points3D[:,1], points3D[:,2], c=color3D/255, cmap="rgb")
ax2.view_init(elev=-60, azim=-90)

fig.show();
```
