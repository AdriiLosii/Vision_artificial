# 3.1 Homografas 2D

Unha homografa 2D  unha transformacin lineal, moi xeral, entre planos. No caderno anterior aplicamos rotacins e translacins a puntos 3D, que son exemplos de homografas euclidianas 3D.

En visin artificial, unha homografa adoita definirse como unha **transformacin de perspectiva dun plano**, noutras verbas, unha reproxeccin dun plano imaxe dunha cmara a outra diferente, onde a cmara puido ser trasladada e/ou rotada. Como consecuencia, **das imaxes calquera dunha mesma superficie plana no espazo estn relacionadas mediante unha homografa**. Isto ten moitas aplicacins prcticas, como a rectificacin e rexistrado de imaxes, realidade aumentada, etc.

Para as homografas 2D, temos unha matriz $3x3$ que contn a transformacin lineal:$\\[010pt]$


$$
\begin{bmatrix} a & b & c \\ d & e & f \\ g & h & 1 \end{bmatrix} \begin{bmatrix} x \\ y \\  1\end{bmatrix} 
=  
\begin{bmatrix} u \\ v \\  w\end{bmatrix}\hspace{2cm} \begin{array}{rcr} x' = u\ /\ w \\ y' = v \ / \ w \end{array}\\[5pt] $$ 

<center>
    <img src="./images/intro_homography.png" width="400"/>
</center>$\\[5pt]$

Ten en conta que nestas transformacins:
- As lias mantense rectas.
- Consrvanse as lias incidentes.

### <span style="color:orange">NIVEL 3: Similaridade/semellanza</span>

A transformacin de semellanza engade un escalado  transforamcin anterior, polo que:

$$\lambda \begin{bmatrix}x' \\ y' \\ 1 \end{bmatrix} = \begin{bmatrix} s & 0 & 0 \\ 0 & s & 0 \\ 0 & 0 & 1 \end{bmatrix}\begin{bmatrix} cos(\theta) & -sin(\theta) & t_x \\ sin(\theta) & cos(\theta) & t_y \\ 0 & 0 & 1 \end{bmatrix}\begin{bmatrix}x \\ y \\ 1 \end{bmatrix} = \begin{bmatrix} s\cdot cos(\theta) & -s\cdot sin(\theta) & s\cdot t_x \\ s\cdot sin(\theta) & s\cdot cos(\theta) & s\cdot t_y \\ 0 & 0 & 1 \end{bmatrix}\begin{bmatrix}x \\ y \\ 1 \end{bmatrix} = \lambda \begin{bmatrix}s\cdot (cos(\theta)x - sin(\theta)y + t_x) \\ s\cdot (sin(\theta)x + cos(\theta)y  + t_y) \\ 1 \end{bmatrix}\\[5pt]$$

 dicir, esta homoografa aplcase **rotacin + traslacin + escala**.
*Ten en conta que a escala ten que ser igual para os dous eixes $x$ e $y$.*

## 3.2 Resolucin da homografa 2D: transformacin lineal directa (DLT) <a id="32"></a>

Como sabedes, unha homografa 2D xenrica que converte un punto $p=(x,y)$ noutro punto $p'=(x',y')$ ten o seguinte aspecto (en coordenadas homoxneas):$\\[5pt ]$

$$\lambda \begin{bmatrix}x' \\ y' \\ 1 \end{bmatrix} = \begin{bmatrix} h_{00} & h_{01} & h_{02} \\ h_{10} & h_{11} & h_{12}\\ h_{20} & h_{21} & 1 \end{bmatrix}\begin{bmatrix}x \\ y \\ 1 \end{bmatrix}$$


Se coecemos as coordenadas dos puntos correspondentes en ambas imaxes, podemos construr un sistema de ecuacins para resolver os coeficientes da matriz. Ademais, coecer unha certa cantidade de puntos correspondentes que comparten a mesma homografa permitiranos obter un sistema de ecuacins (sobre)determinado para obtelos.

Neste caso, a partir dun determinado par de puntos $i$, podemos formar un sistema de ecuacins lineais para obter as variables da matriz:$\\[5pt]$


$$\begin{eqnarray}
x'_i = \frac{h_{00}x_i + h_{01}y_i + h_{02}}{h_{20}x_i + h_{21}y_i + 1}\  \longrightarrow \ x'_i(h_{20}x_i + h_{21}y_i + 1) = h_{00}x_i + h_{01}y_i + h_{02} \\[10pt]
y'_i = \frac{h_{10}x_i + h_{11}y_i + h_{12}}{h_{20}x_i + h_{21}y_i + 1}\ \longrightarrow \ y'_i(h_{20}x_i + h_{21}y_i + 1) = h_{10}x_i + h_{11}y_i + h_{12}
\end{eqnarray}
$$
$\\[10pt]$

Isto pode ser reescrito como: 

$$A\mathbf{h} = 0 \longrightarrow
\begin{bmatrix}
-x_i & -y_i & -1 & 0 & 0 & 0 & x'_ix_i & x'_iy_i & x'_i \\
0 & 0 & 0 & -x_i & -y_i & -1 & y'_ix_i & y'_iy_i & y'_i
\end{bmatrix} 
\begin{bmatrix}
h_{00} \\ h_{01} \\ h_{02} \\ h_{10} \\ h_{11} \\ h_{12} \\ h_{20} \\ h_{21} \\ 1
\end{bmatrix} 
=
\begin{bmatrix}
0 \\ 0 
\end{bmatrix} 
\\[5pt]
$$

Podes ver que con menos de 4 puntos independentes ($rank(A) \lt 8$), hai infinitas solucins detrs de $k\textbf{h}$.

Pero, se temos **4 ou mis pares de puntos independentes** podemos construr a seguinte expresin:$\\[5pt]$


$$Ah = 0 \longrightarrow
\begin{bmatrix}
-x_1 & -y_1 & -1 & 0 & 0 & 0 & x'_1x_1 & x'_1y_1 & x'_1 \\
0 & 0 & 0 & -x_1 & -y_1 & -1 & y'_1x_1 & y'_1y_1 & y'_1 \\
& & & &\vdots \\
-x_n & -y_n & -1 & 0 & 0 & 0 & x'_nx_n & x'_ny_n & x'_n \\
0 & 0 & 0 & -x_n & -y_n & -1 & y'_nx_n & y'_ny_n & y'_n \\
\end{bmatrix} 
\begin{bmatrix}
h_{00} \\ h_{01} \\ h_{02} \\ h_{10} \\ h_{11} \\ h_{12} \\ h_{20} \\ h_{21} \\ 1
\end{bmatrix} 
= 
\begin{bmatrix}
0 \\0\\ \vdots \\ 0 \\ 0
\end{bmatrix} 
$$


- Para $\textbf{n = 4}$ **pares de puntos independentes** ($rank(A) = 8$), hai unha solucin $k\textbf{h}$ con $\textbf{h} \ne 0, k \ne 0$. Isto debera ser suficiente (e en rigor o , como probaremos mis adiante), pero a solucin verase moi afectada polo rudo nas coordenadas dos puntos, polo que non se adoita empregar en aplicacins con puntos detectados automaticamente, senn para aqueles elixidos manualmente. $\\[5pt]$
- Para $\textbf{n > 4}$ **pares de puntos independentes** ($rank(A) \gt 8$), (por exemplo, aqueles pares de puntos detectados por SIFT ou Harris) non hai solucin (aparte de $\textbf{h} = 0$) xa que o sistema est sobredeterminado. Non obstante, podemos obter a solucin que minimiza o erro nas coordenadas dos puntos ( dicir, a solucin de mnimos cadrados): $\\[5pt]$

    - Arg. min$_h ||\textbf{Ah}||^2$ con $||\textbf{h}|| = 1$$\\[5pt]$
    - Solucin $\hat{h}$: vector propio do menor valor propio de $A^TA$
    
Tea en conta que se coece o tipo especfico de transformacin, pode necesitar menos puntos para resolver este sistema de ecuacins:

- **Translacin** (2 incgnitas) $\rightarrow$ **Necestase 1 par de puntos**
- **Euclidiana** (3 incgnitas) $\rightarrow$ **Necestanse 2 pares de puntos**
- **Semellanza** (4 incgnitas) $\rightarrow$ **Necestanse 2 pares de puntos**
- **Afn** (6 incgnitas) $\rightarrow$ **Necestanse 3 pares de puntos**
- **Proxectiva** (8 incgnitas) $\rightarrow$ **Necestanse 4 pares de puntos**

Pero, como dixemos, normalmente, necesitas moitos mis puntos para a robustez do rudo, polo que isto non  relevante.

###  hora de inclur as lias

Finalmente,  hora de **debuxar as lias**. Deberas ter gardado as coordenadas dos dous puntos para cada lia da imaxe sen perspectiva.

Para a homografa de lias, s tes que **aplicar a homografa a dous puntos da lia**. Deste xeito, ters dous puntos na imaxe orixinal que definen a lia con perpectiva.

Tea en conta que `cv2.line()` ten dous puntos como entrada, pero s debuxa a lia do segmento. Ters que **estender ese segmento** usando a pendente da recta ata que cruce todo o campo de xogo.


$$pendente = \frac{y_2-y_1}{x_2-x_1}$$

Os novos puntos sern:

$$\begin{eqnarray}
x'_1,\ \  y'_1 = x_1 + extension,\ \ y_1 + extension\times pendente\\[3pt]
x'_2,\ \  y'_2 = x_2 - extension,\ \  y_2 - extension\times pendente
\end{eqnarray}$$

### **<span style="color:green"><b><i>TAREFA 2d: Debuxando as lias</i></b></span>**

Aplica a **homografa inversa** aos puntos da lia, **estende** o segmento de lia e **debuxa** a lia de inicio (azul) e a lia de down (amarela). Finalmente, visualiza a imaxe resultante.

### Unha segunda forma cando se trata de lias: empregando a sa ecuacin

Unha forma alternativa de aplicar unha homoografa a unha lia  considerando a sa ecuacin:

$$ax+by+c=0$$

Logo a lia  $l=[a,b,c]^T$. Deste xeito, unha lia vertical na imaxe sen proxeccin estara representada por un vector da forma $l'=[1, 0, -x]^T$, polo que s necesitamos a coordenada $x$ (columna) para construr esta lia!

Unha vez que dispoemos da lia $l'$, a mesma homografa utilizada para transformar puntos pdese utilizar aqu para obter o seu equivalente na imaxe con proxeccin:

$$l = H^T l'$$

Incrible, non? Btalle un ollo ao seguinte cdigo que realiza estes clculos e obters os mesmos resultados que antes.

```python
pers_with_lines = perspective.copy()

x1 = start_line[0,0]

l_h = np.vstack(np.array([1,0,-x1]))

l = M.T@l_h
n_y1 = 0
n_y2 = pers_with_lines.shape[0]
n_x1 = (-l[1]*n_y1-l[2])/l[0]
n_x2 = (-l[1]*n_y2-l[2])/l[0]
pers_with_lines = cv2.line(pers_with_lines, (n_x1,n_y1), (n_x2,n_y2),thickness=10,color=[0,0,155])

x1 = down_line[0,0]

l_h = np.vstack(np.array([1,0,-x1]))

l = M.T@l_h
n_y1 = 0
n_y2 = perspective.shape[0]
n_x1 = (-l[1]*n_y1-l[2])/l[0]
n_x2 = (-l[1]*n_y2-l[2])/l[0]
pers_with_lines = cv2.line(pers_with_lines, (n_x1,n_y1), (n_x2,n_y2),thickness=10,color=[255,255,0])

plt.imshow(pers_with_lines);
```
